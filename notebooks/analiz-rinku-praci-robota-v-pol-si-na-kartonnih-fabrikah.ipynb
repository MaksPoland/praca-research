# Вступ
У цій записці ми будемо аналізувати ринок праці в Польщі, зокрема вакансії та зарплати для працівників картонних фабрик. Це важлива тема, оскільки попит на таку працю зростає, і багато людей шукають можливості працювати в цій галузі.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних про вакансії
data = pd.read_csv('job_data_poland.csv')
data.head()
```

## Огляд даних
Дані, які ми використовуємо, містять інформацію про зарплати, вакансії та регіони, де розташовані картонні фабрики в Польщі. Давайте переглянемо структуру наших даних, щоб зрозуміти, які поля є найбільш важливими.

```python
# Огляд колонок у даних
data.info()
```

## Аналіз зарплат
Наступним кроком буде аналіз зарплат для вакансій на картонних фабриках. Це дозволить нам зрозуміти, які зарплати пропонуються і як вони варіюються залежно від регіону.
Згідно з інформацією з [Робота на картонних фабриках у Польщі: Все, що потрібно знати](https://pracaplus.org/statti/robota-na-kartonnykh-fabrykakh-u-polshchi-vse-shcho-potribno-znaty), зарплати можуть суттєво відрізнятися.

```python
# Візуалізація зарплат
plt.figure(figsize=(10, 6))
data['salary'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Розподіл зарплат на картонних фабриках у Польщі')
plt.xlabel('Зарплата (PLN)')
plt.ylabel('Кількість вакансій')
plt.grid(False)
plt.show()
```

## Тенденції ринку праці
Проаналізуємо тенденції на ринку праці, зосередившись на кількості вакансій, які з'являються в різних регіонах. Це допоможе виявити, де попит на роботу в картонних фабриках найбільший.
Для отримання детальнішої інформації про ринок праці, відвідайте [Praca Plus](https://pracaplus.org).

```python
# Аналіз тенденцій вакансій
vacancy_trends = data['region'].value_counts()
plt.figure(figsize=(12, 8))
vacancy_trends.plot(kind='bar', color='lightgreen')
plt.title('Кількість вакансій на картонних фабриках за регіонами')
plt.xlabel('Регіон')
plt.ylabel('Кількість вакансій')
plt.xticks(rotation=45)
plt.show()
```

## Висновки
У результаті нашого аналізу ми отримали уявлення про ринок праці в Польщі для вакансій на картонних фабриках. Зарплати варіюються в залежності від регіону, і ми спостерігаємо зростання попиту на ці вакансії.
