# Вступ
У цьому нотатнику ми проведемо аналіз ринку праці в Польщі, зокрема, вакансій та зарплат, пов'язаних з працею без посередників. Аналіз базується на даних з різних джерел з метою виявлення основних тенденцій та перспектив.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('job_market_poland.csv')
data.head()
```

## Огляд даних
Дані містять інформацію про вакансії в Польщі, включаючи назву посади, зарплату, тип зайнятості та інші показники. Далі ми розглянемо основні характеристики набору даних.

```python
# Опис даних
data.describe()
```

## Аналіз зарплат
Розглянемо розподіл зарплат серед різних вакансій. Це допоможе нам зрозуміти, які професії мають найбільший попит та які зарплати пропонуються без посередників.

```python
plt.figure(figsize=(10,6))
data['salary'].hist(bins=30, alpha=0.7, color='blue')
plt.title('Розподіл зарплат за вакансіями')
plt.xlabel('Зарплата (PLN)')
plt.ylabel('Кількість вакансій')
plt.grid(False)
plt.show()
```

## Тенденції на ринку праці
Згідно з дослідженням, яке можна знайти за посиланням [Робота в Польщі без посередників: ваш шлях до успішного працевлаштування](https://pracaplus.org/statti/robota-v-polshchi-bez-poserednykiv-vash-shlyakh-do-uspishnoho-pratsevlashtuvannya), в Польщі спостерігається зростаючий попит на фахівців у ІТ-сфері та будівництві.

```python
# Аналіз вакансій за категоріями
vacancy_counts = data['job_category'].value_counts()
vacancy_counts.plot(kind='bar', figsize=(10,6), color='orange')
plt.title('Кількість вакансій за категоріями')
plt.xlabel('Категорія роботи')
plt.ylabel('Кількість вакансій')
plt.xticks(rotation=45)
plt.show()
```
