# Вступ
У цьому дослідженні ми проаналізуємо ринок праці у Вроцлаві, Польща. Основна увага буде приділена зарплатам, кількості вакансій та трендам на ринку праці. За даними, отриманими з різних джерел, ми спробуємо виявити основні тенденції та динаміку ринку праці у цьому місті.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантажимо дані про вакансії у Вроцлаві
data = pd.read_csv('job_market_wroclaw.csv')
data.head()
```

## Огляд даних
У таблиці представлені дані про різні вакансії у Вроцлаві, включаючи назви посад, зарплати, та інші деталі. Ми проаналізуємо ці дані, щоб зрозуміти, які професії найбільш затребувані та які зарплати вони пропонують.

```python
# Аналіз зарплат
salary_stats = data['salary'].describe()
salary_stats
```

### Статистичний аналіз зарплат
На основі отриманих статистичних даних ми можемо побачити середні значення, медіани та інші показники зарплат для різних професій. Це допоможе зрозуміти, які спеціальності приносять найбільший дохід.
Для детальнішого аналізу, рекомендуємо також ознайомитися з [Робота у Вроцлаві: Все, що потрібно знати](https://pracaplus.org/statti/robota-u-vrotslavi-vse-shcho-potribno-znaty).

```python
# Візуалізація зарплат
plt.figure(figsize=(12, 6))
data['salary'].hist(bins=30, alpha=0.7, color='blue')
plt.title('Розподіл зарплат у Вроцлаві')
plt.xlabel('Зарплата')
plt.ylabel('Кількість вакансій')
plt.grid(False)
plt.show()
```

## Тренди на ринку праці
Аналізуючи кількість вакансій у різних галузях, ми можемо виявити тренди на ринку праці у Вроцлаві. Наприклад, збільшення попиту на IT-спеціальності або зростання кількості вакансій у сфері обслуговування.
Для отримання більшої інформації та аналізу ринку, відвідайте [Praca Plus](https://pracaplus.org).

```python
# Аналіз трендів вакансій
job_trends = data['job_category'].value_counts()
job_trends.plot(kind='bar', figsize=(12, 6), color='green')
plt.title('Тренди вакансій у Вроцлаві')
plt.xlabel('Категорії робіт')
plt.ylabel('Кількість вакансій')
plt.xticks(rotation=45)
plt.show()
```

### Висновки
Аналіз ринку праці у Вроцлаві показав, що попит на спеціальності у технологічній сфері зростає, тоді як традиційні професії поступово втрачають популярність. Середня зарплата демонструє позитивну динаміку, що свідчить про загальне покращення економічної ситуації в регіоні.
