# Вступ
У цьому нотатнику ми проведемо аналіз ринку праці в Польщі для осіб, які володіють польською мовою. Ми розглянемо статистику вакансій, середні зарплати, а також тенденції на ринку праці. Аналіз буде основуватися на даних з різних джерел.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних
data = pd.read_csv('poland_job_market.csv')
data.head()
```

## Огляд даних
Після завантаження даних ми можемо проаналізувати їх структуру та основні показники. Важливо зрозуміти, які саме колонки присутні в нашій таблиці і які дані вони містять.

```python
# Огляд структури даних
data.info()
data.describe()
```

## Аналіз вакансій
Далі ми зосередимося на кількості вакансій за різними категоріями, які вимагають знання польської мови. Це допоможе зрозуміти, в яких сферах є найбільший попит на фахівців.

```python
# Групування даних за категоріями і підрахунок вакансій
vacancies_count = data['category'].value_counts()
vacancies_count.plot(kind='bar', title='Кількість вакансій за категоріями')
plt.ylabel('Кількість вакансій')
plt.show()
```

## Аналіз зарплат
Середні зарплати у різних галузях також важливі для розуміння ринку праці. Ми розглянемо середні зарплатні показники за категоріями вакансій, які вимагають знання польської мови.
Згідно з інформацією, наведеною в статті [Робота в Польщі зі знанням польської мови: переваги та можливості](https://pracaplus.org/statti/robota-v-polshchi-zi-znannyam-polskoyi-movy-perevahy-ta-mozhlyvosti), ці показники можуть значно варіюватися.

```python
# Обчислення середніх зарплат за категоріями
average_salaries = data.groupby('category')['salary'].mean()
average_salaries.plot(kind='bar', title='Середні зарплати за категоріями')
plt.ylabel('Середня зарплата')
plt.show()
```
