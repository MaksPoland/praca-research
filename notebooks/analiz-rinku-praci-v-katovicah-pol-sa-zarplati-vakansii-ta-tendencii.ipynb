# Вступ
У цьому дослідженні ми розглянемо ринок праці в Катовіцах, Польща. 
Ми проаналізуємо дані про зарплати, вакансії та тенденції, щоб зрозуміти, які можливості є для українців, які шукають роботу в цьому регіоні.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних про вакансії
data = pd.read_csv('katowice_job_data.csv')
data.head()
```

## Аналіз даних
Після завантаження даних ми можемо провести початковий аналіз. 
Ці дані містять інформацію про типи вакансій, зарплати та інші важливі показники.

```python
# Огляд статистики даних
data.describe()
```

## Візуалізація зарплат
Тепер ми можемо візуалізувати розподіл зарплат на ринку праці в Катовіцах.
Це допоможе зрозуміти, які спеціальності надають найвищі можливості для заробітку.

```python
plt.figure(figsize=(10,6))
data['salary'].hist(bins=30, edgecolor='black')
plt.title('Розподіл зарплат у Катовіцах')
plt.xlabel('Зарплата')
plt.ylabel('Кількість вакансій')
plt.grid(False)
plt.show()
```

## Тенденції на ринку праці
Згідно з даними, можна зазначити, що у Катовіцах спостерігається зростання кількості вакансій у сфері IT та інженерії. 
Це підтверджує інформацію з [Робота в Катовіцах: можливості та поради для українців](https://pracaplus.org/statti/robota-v-katovitsakh-mozhlyvosti-ta-porady-dlya-ukrayintsiv).

```python
# Групування вакансій за типом роботи
job_types = data['job_type'].value_counts()
job_types.plot(kind='bar', figsize=(10, 5))
plt.title('Кількість вакансій за типом роботи')
plt.xlabel('Тип роботи')
plt.ylabel('Кількість вакансій')
plt.xticks(rotation=45)
plt.show()
```
