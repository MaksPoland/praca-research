# Аналіз ринку праці в Польщі
Цей ноутбук присвячений аналізу ринку праці в Польщі, зокрема вакансій та зарплат. Ми дослідимо тенденції, які стосуються пошуку роботи в Польщі, а також середні зарплати в різних секторах.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('job_market_poland.csv')
data.head()
```

## Огляд даних
Після завантаження даних ми можемо переглянути перші кілька рядків, щоб зрозуміти, які інформаційні поля містять наші дані. Це допоможе нам ідентифікувати основні фактори, які впливають на ринок праці.

```python
# Огляд статистики даних
data.describe()
```

## Аналіз заробітних плат
Зарплати є ключовим фактором при виборі роботи. Давайте візуалізуємо середні зарплати в різних секторах, щоб зрозуміти, які галузі пропонують найкращі умови праці.

```python
plt.figure(figsize=(10, 6))
salary_by_sector = data.groupby('sector')['average_salary'].mean().sort_values()
salary_by_sector.plot(kind='barh', color='skyblue')
plt.title('Середні зарплати за секторами')
plt.xlabel('Середня зарплата')
plt.ylabel('Сектор')
plt.show()
```

## Тенденції ринку праці
Аналізуючи дані про вакансії, ми можемо виявити тенденції, які можуть допомогти шукачам роботи зрозуміти, які професії є найбільш затребуваними. Це також корисно для підприємств, які планують наймання нових співробітників.
Додаткову інформацію про те, як знайти легальну роботу в Польщі, можна знайти за посиланням: [Вакансії в Польщі: Як знайти легальну роботу](https://pracaplus.org/statti/vakansiyi-v-polshchi-yak-znayty-lehalnu-robotu).

```python
# Аналіз вакансій
vacancies_count = data['job_title'].value_counts().head(10)
vacancies_count.plot(kind='bar', color='orange')
plt.title('Топ-10 вакансій у Польщі')
plt.xlabel('Вакансії')
plt.ylabel('Кількість')
plt.show()
```
