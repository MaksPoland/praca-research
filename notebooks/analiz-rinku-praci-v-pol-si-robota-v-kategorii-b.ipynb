# Вступ
У цьому нотатнику ми проведемо аналіз ринку праці в Польщі, зосередившись на вакансіях категорії Б. Категорія Б охоплює різноманітні професії, які вимагають спеціальної підготовки та дозволів. Ми розглянемо дані про зарплати, кількість вакансій та тенденції на ринку праці.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних
data = pd.read_csv('job_market_poland_category_b.csv')
data.head()
```

## Огляд даних
Перш ніж почати аналіз, розглянемо структуру даних, які ми завантажили. В даних містяться такі стовпці, як: ' вакансія', ' зарплата', ' місто', ' дата публікації'. Це дозволить нам краще зрозуміти ринок праці.

```python
# Статистичний опис даних
data.describe()
```

## Аналіз зарплат
Важливо зрозуміти, як зарплати варіюються залежно від регіону та типу вакансій. Давайте побудуємо графік зарплат для різних міст.

```python
plt.figure(figsize=(12, 6))
data.groupby('місто')['зарплата'].mean().sort_values().plot(kind='barh')
plt.title('Середня зарплата за містами для категорії Б')
plt.xlabel('Середня зарплата')
plt.ylabel('Місто')
plt.show()
```

## Тенденції на ринку праці
Зараз розглянемо, як змінювалась кількість вакансій з часом. Це допоможе зрозуміти, чи зростає попит на працівників категорії Б в Польщі.

```python
data['дата публікації'] = pd.to_datetime(data['дата публікації'])
vacancies_trend = data.groupby(data['дата публікації'].dt.to_period('M')).size()
plt.figure(figsize=(12, 6))
vacancies_trend.plot()
plt.title('Тенденції вакансій категорії Б в Польщі')
plt.xlabel('Дата')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```

## Висновки
Аналіз показав, що ринок праці для категорії Б в Польщі активно розвивається. Середня зарплата варіюється залежно від міста, а також ми спостерігаємо зростання кількості вакансій в останні роки.
