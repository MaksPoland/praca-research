# Вступ
Цей ноутбук аналізує ринок праці в Польщі для студентів. Особливу увагу буде приділено можливостям працевлаштування, зарплатам та тенденціям на ринку. Ми використаємо дані про вакансії, щоб зрозуміти, які професії є найбільш затребуваними для студентів.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних про вакансії
data = pd.read_csv('job_market_data.csv')
data.head()
```

# Огляд даних
У цьому розділі ми розглянемо структуру вхідних даних. Маючи інформацію про зарплати та вакансії, ми можемо зробити висновки про стан ринку праці для студентів у Польщі.

```python
# Опис даних
data.info()

# Визначення кількості вакансій по професіях
vacancies_count = data['profession'].value_counts()
vacancies_count.plot(kind='bar', title='Кількість вакансій по професіях')
plt.xlabel('Професії')
plt.ylabel('Кількість вакансій')
plt.show()
```

# Тенденції на ринку праці
Аналізуючи дані, ми можемо виявити тенденції в зарплатах та кількості вакансій. Це дозволяє студентам краще орієнтуватися у виборі спеціальності.
Згідно з [Робота в Польщі для студентів: можливості та поради](https://pracaplus.org/statti/robota-v-polshchi-dlya-studentiv-mozhlyvosti-ta-porady), ринок праці в Польщі активно розвивається.

```python
# Візуалізація зарплат по професіях
salary_by_profession = data.groupby('profession')['salary'].mean().sort_values()
salary_by_profession.plot(kind='barh', title='Середня зарплата по професіях')
plt.xlabel('Середня зарплата')
plt.ylabel('Професії')
plt.show()
```
