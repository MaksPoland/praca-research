У цьому дослідженні ми будемо аналізувати ринок праці в Польщі для жінок, які не володіють мовою. Зокрема, ми зосередимося на кількості вакансій, середній заробітній платі та загальних тенденціях на ринку. Це важлива тема, оскільки багато жінок шукають можливості трудової діяльності в Польщі без знання польської мови.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних про вакансії
data = pd.read_csv('job_market_data.csv')
data.head()
```

## Огляд даних
Ми завантажили дані, які містять інформацію про вакансії в Польщі для жінок без знання мови. Далі ми розглянемо структуру даних та ключові показники.

```python
# Огляд стовпців даних
data.info()

# Опис статистичних характеристик
data.describe()
```

## Аналіз заробітної плати
На основі отриманих даних, ми проаналізуємо середній рівень заробітної плати для різних вакансій. Це дозволить нам зрозуміти можливості, які можуть бути доступні жінкам без знання мови.

```python
# Групування даних за типом вакансії та обчислення середньої заробітної плати
salary_analysis = data.groupby('job_type')['salary'].mean().sort_values()
salary_analysis.plot(kind='barh', title='Середня заробітна плата за типом вакансії', xlabel='Зарплата', ylabel='Тип вакансії')
plt.show()
```
