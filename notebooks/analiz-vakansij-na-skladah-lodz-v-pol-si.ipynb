# Аналіз вакансій на складах Лодзь в Польщі
Цей нотатник присвячений аналізу ринку праці, зосередженому на вакансіях у складському секторі Лодзі. Ми розглянемо дані про зарплати, кількість вакансій та тенденції на цьому ринку.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('lodz_warehouse_jobs.csv')
data.head()
```

## Огляд даних
В даному наборі даних ми маємо інформацію про вакансії на складах Лодзь, включаючи зарплати, типи вакансій та дати публікацій. Проведемо попередній аналіз, щоб зрозуміти структуру даних.

```python
# Огляд основних статистичних характеристик
data.describe()
```

## Аналіз зарплат
Зараз ми проаналізуємо розподіл зарплат для різних вакансій. Зокрема, ми хочемо визначити середню зарплату та візуалізувати її розподіл.

```python
# Аналіз зарплат
plt.figure(figsize=(10, 6))
plt.hist(data['salary'], bins=20, color='blue', alpha=0.7) 
plt.title('Розподіл зарплат для вакансій на складах Лодзь')
plt.xlabel('Зарплата (PLN)')
plt.ylabel('Кількість вакансій')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Тенденції на ринку вакансій
Аналізуючи дані про публікації вакансій за часом, ми можемо виявити тенденції. Це допоможе зрозуміти, чи зростає попит на працівників у складському секторі.

```python
# Тенденції вакансій
data['date'] = pd.to_datetime(data['date'])
vacancies_per_month = data.groupby(data['date'].dt.to_period('M')).size()
vacancies_per_month.plot(kind='line', figsize=(10, 6), title='Кількість вакансій на складах Лодзь за місяцями')
plt.xlabel('Місяць')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```

Як видно з графіка, є чіткі тенденції у кількості вакансій. Це може свідчити про сезонність попиту на працівників.
