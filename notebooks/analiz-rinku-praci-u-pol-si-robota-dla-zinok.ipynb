# Вступ
У цьому нотатнику ми проаналізуємо ринок праці у Польщі, з особливим акцентом на можливості та виклики для жінок. Згідно з дослідженнями, жінки все частіше займають активну роль у професійному середовищі, проте існують різні аспекти, які впливають на їх успіх. В даному аналізі ми розглянемо статистику вакансій, рівні заробітної плати та загальні тенденції на ринку.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('poland_job_market.csv')
data.head()
```

## Огляд даних
Ми почнемо з огляду даних, що містять інформацію про вакансії, рівні заробітної плати та інші релевантні показники. Це дозволить нам зрозуміти основні характеристики ринку праці для жінок у Польщі.

```python
# Огляд статистики вакансій і заробітних плат
summary = data.describe()
summary[['salary', 'vacancies']].plot(kind='bar')
plt.title('Середня заробітна плата та кількість вакансій')
plt.xlabel('Категорії')
plt.ylabel('Значення')
plt.show()
```

## Аналіз тенденцій
Аналізуючи тенденції, ми можемо зазначити, що деякі галузі, такі як IT та охорона здоров'я, пропонують більше можливостей для жінок. Наприклад, жінки, які працюють у технологічному секторі, часто мають вищі заробітки порівняно з іншими сферами.
Також важливо врахувати культурні та соціальні фактори, які можуть впливати на вибір професії.
Джерело: [Польща: Робота для жінок — можливості та виклики](https://pracaplus.org/statti/polshcha-robota-dlya-zhinok-mozhlyvosti-ta-vyklyky)

```python
# Аналіз вакансій за галузями
industry_counts = data['industry'].value_counts()
industry_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Розподіл вакансій за галузями')
plt.ylabel('')
plt.show()
```
