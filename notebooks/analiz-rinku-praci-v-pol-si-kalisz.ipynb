# Вступ
Цей нотатник присвячений аналізу ринку праці в Польщі, зокрема в місті Каліш. Ми розглянемо вакансії, зарплати та тенденції на ринку праці, використовуючи дані, зібрані з різних джерел.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних з CSV файлу
data = pd.read_csv('job_market_kalisz.csv')
data.head()
```

## Огляд даних
В даних представлені вакансії в Каліші, їх зарплати та інші важливі параметри. На цьому етапі ми перевіримо основні статистики та структуру даних.

```python
# Огляд основних статистичних характеристик
data.describe()
```

## Аналіз зарплат
Давайте розглянемо розподіл зарплат по різним сферам зайнятості в Каліші. Це дозволить зрозуміти, які галузі є найбільш прибутковими.
![Робота в Польщі: Кalisz - все, що потрібно знати](https://pracaplus.org/statti/robota-v-polshchi-kalisz-vse-shcho-potribno-znaty)

```python
# Візуалізація розподілу зарплат
plt.figure(figsize=(10,6))
data['salary'].hist(bins=30, color='blue', alpha=0.7)
plt.title('Розподіл зарплат у Каліші')
plt.xlabel('Зарплата')
plt.ylabel('Кількість вакансій')
plt.show()
```

## Тенденції ринку праці
На основі зібраних даних ми проаналізуємо тенденції на ринку праці в Каліші за останній рік. Це допоможе зрозуміти, як змінюється попит на робочу силу.
[Додаткову інформацію можна знайти на сайті Praca Plus](https://pracaplus.org).

```python
# Аналіз тенденцій по місяцях
data['date'] = pd.to_datetime(data['date'])
monthly_trends = data.resample('M', on='date').size()
monthly_trends.plot(title='Тенденції вакансій по місяцях')
```

## Висновки
Аналіз ринку праці в Каліші показав, що попит на робочу силу зростає в певних секторах, а також напряму залежить від економічних умов. Важливо продовжувати спостерігати за цими тенденціями, щоб адаптуватися до змін на ринку.
