# Аналіз ринку праці в Польщі
У цьому ноутбуці ми проведемо аналіз ринку праці у сфері прибирання на дому. Зокрема, ми розглянемо дані щодо заробітних плат, вакансій та тенденцій у цій галузі.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('cleaning_jobs_poland.csv')
data.head()
```

## Огляд даних
Ми завантажили дані про вакансії на роботу прибирання на дому в Польщі. У цьому наборі даних містяться інформація про заробітну плату, кількість вакансій та регіони.

```python
# Опис даних
data.describe()
```

## Аналіз заробітних плат
Далі ми проаналізуємо, які середні заробітні плати пропонуються для вакансій з прибирання на дому в Польщі.

```python
# Аналіз середньої заробітної плати
average_salary = data['salary'].mean()
plt.figure(figsize=(10, 6))
plt.title('Середня заробітна плата у прибиранні на дому в Польщі')
plt.bar(['Середня зарплата'], [average_salary], color='blue')
plt.ylabel('Заробітна плата (PLN)')
plt.show()
```

### Тенденції на ринку праці
Згідно з даними, ми можемо спостерігати певні тенденції на ринку праці для робітників у сфері прибирання на дому. Основні фактори, які впливають на попит і пропозицію, включають сезонність та економічну ситуацію.
Додаткову інформацію про ринок праці в цій галузі можна знайти в статті: [Робота прибирання на дому в Польщі: Все, що потрібно знати](https://pracaplus.org/statti/robota-prybyrannya-na-domu-v-polshchi-vse-shcho-potribno-znaty).

```python
# Аналіз вакансій за регіонами
vacancies_by_region = data.groupby('region')['vacancies'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 7))
vacancies_by_region.plot(kind='bar', color='orange')
plt.title('Кількість вакансій у прибиранні на дому за регіонами')
plt.xlabel('Регіон')
plt.ylabel('Кількість вакансій')
plt.xticks(rotation=45)
plt.show()
```
