# Аналіз ринку праці в Польщі
Цей ноутбук присвячений аналізу ринку праці в Польщі, зокрема темі податків для українців, таких як PIT та ZUS. Ми розглянемо дані щодо зарплат, вакансій та тенденцій на ринку праці.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('job_market_data.csv')
data.head()
```

## Огляд даних
На цьому етапі ми підготували дані для аналізу. Розглянемо основні стовпці нашого датасету, які містять інформацію про зарплати та кількість вакансій.

```python
# Опис структури даних
data.info()
```

## Аналіз зарплат
Важливо зрозуміти, які зарплати пропонуються на ринку праці для українців. Пропонуємо побудувати графік, що показує середні зарплати за різними професіями.

```python
# Групування даних за професіями
average_salaries = data.groupby('profession')['salary'].mean().sort_values()
plt.figure(figsize=(10,6))
average_salaries.plot(kind='barh')
plt.title('Середні зарплати за професіями')
plt.xlabel('Зарплата (PLN)')
plt.ylabel('Професії')
plt.show()
```

## Тенденції на ринку праці
Окрім зарплат, важливо аналізувати тенденції вакансій. Дослідимо, які професії найбільш затребувані в Польщі для українців.

```python
# Групування даних за кількістю вакансій
vacancies_count = data['profession'].value_counts()
plt.figure(figsize=(10,6))
vacancies_count.plot(kind='bar')
plt.title('Кількість вакансій за професіями')
plt.xlabel('Професії')
plt.ylabel('Кількість вакансій')
plt.xticks(rotation=45)
plt.show()
```

Згідно з даними, багато українців цікавляться роботою в Польщі, і важливо враховувати податки, такі як PIT та ZUS. Більш детально про податки можна прочитати у статті [Податки в Польщі для українців: PIT та ZUS простими словами](https://pracaplus.org/statti/podatky-v-polshchi-dlya-ukrayintsiv-pit-ta-zus-prostymy-slovamy).
