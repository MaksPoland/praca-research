# –í—Å—Ç—É–ø
–£ —Ü—å–æ–º—É –¥–æ—Å–ª—ñ–¥–∂–µ–Ω–Ω—ñ –º–∏ –∞–Ω–∞–ª—ñ–∑—É—î–º–æ —Ä–∏–Ω–æ–∫ –ø—Ä–∞—Ü—ñ –≤ –ü–æ–ª—å—â—ñ, –∑–æ–∫—Ä–µ–º–∞ –º–æ–∂–ª–∏–≤–æ—Å—Ç—ñ –¥–ª—è –∂—ñ–Ω–æ–∫, —è–∫—ñ –ø—Ä–∞—Ü—é—é—Ç—å –∑–∞ –±—ñ–æ–º–µ—Ç—Ä–∏—á–Ω–∏–º–∏ –ø–∞—Å–ø–æ—Ä—Ç–∞–º–∏. –ó—Ä–æ—Å—Ç–∞—é—á–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∂—ñ–Ω–æ–∫, —è–∫—ñ —à—É–∫–∞—é—Ç—å —Ä–æ–±–æ—Ç—É –≤ –ü–æ–ª—å—â—ñ, –ø—ñ–¥–∫—Ä–µ—Å–ª—é—î –ø–æ—Ç—Ä–µ–±—É –≤ —Ä–æ–∑—É–º—ñ–Ω–Ω—ñ –∑–∞—Ä–ø–ª–∞—Ç, –≤–∞–∫–∞–Ω—Å—ñ–π —ñ —Ç–µ–Ω–¥–µ–Ω—Ü—ñ–π –Ω–∞ —Ä–∏–Ω–∫—É –ø—Ä–∞—Ü—ñ. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –¥–∞–Ω–∏—Ö –ø—Ä–æ –≤–∞–∫–∞–Ω—Å—ñ—ó
data = pd.read_csv('vacancies_poland.csv')
data.head()
```

## –û–≥–ª—è–¥ –¥–∞–Ω–∏—Ö
–î–∞–Ω—ñ, —è–∫—ñ –º–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–ª–∏, –º—ñ—Å—Ç—è—Ç—å —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—é –ø—Ä–æ –≤–∞–∫–∞–Ω—Å—ñ—ó –≤ –ü–æ–ª—å—â—ñ, –≤–∫–ª—é—á–∞—é—á–∏ –∑–∞—Ä–ø–ª–∞—Ç–∏, –ø–æ–∑–∏—Ü—ñ—ó, —ñ –≤–∏–º–æ–≥–∏. –ü–µ—Ä—à –Ω—ñ–∂ –ø–æ—á–∞—Ç–∏ –∞–Ω–∞–ª—ñ–∑, —Ä–æ–∑–≥–ª—è–Ω–µ–º–æ –æ—Å–Ω–æ–≤–Ω—ñ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –Ω–∞—à–æ–≥–æ –Ω–∞–±–æ—Ä—É –¥–∞–Ω–∏—Ö.

```python
# –û—Å–Ω–æ–≤–Ω—ñ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω—ñ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏
data.describe()
```

## –ê–Ω–∞–ª—ñ–∑ –∑–∞—Ä–ø–ª–∞—Ç
–ù–∞ –æ—Å–Ω–æ–≤—ñ –∑—ñ–±—Ä–∞–Ω–∏—Ö –¥–∞–Ω–∏—Ö –º–∏ –º–æ–∂–µ–º–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ –∞–Ω–∞–ª—ñ–∑ –∑–∞—Ä–ø–ª–∞—Ç. –í–∞–∂–ª–∏–≤–æ –∑—Ä–æ–∑—É–º—ñ—Ç–∏, —è–∫—ñ –ø—Ä–æ—Ñ–µ—Å—ñ—ó –ø—Ä–æ–ø–æ–Ω—É—é—Ç—å –Ω–∞–π–±—ñ–ª—å—à—ñ –∑–∞—Ä–æ–±—ñ—Ç–∫–∏ –¥–ª—è –∂—ñ–Ω–æ–∫, —è–∫—ñ –ø—Ä–∞—Ü—é—é—Ç—å –≤ –ü–æ–ª—å—â—ñ.
–ú–∏ —Ç–∞–∫–æ–∂ –∑–≤–µ—Ä–Ω–µ–º–æ —É–≤–∞–≥—É –Ω–∞ —Ä—ñ–∑–Ω–∏—Ü—é –≤ –∑–∞—Ä–ø–ª–∞—Ç–∞—Ö –º—ñ–∂ —Ä—ñ–∑–Ω–∏–º–∏ —Å—Ñ–µ—Ä–∞–º–∏ –¥—ñ—è–ª—å–Ω–æ—Å—Ç—ñ.

```python
# –ì—Ä—É–ø—É–≤–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö –∑–∞ –ø—Ä–æ—Ñ–µ—Å—ñ—è–º–∏ —Ç–∞ —Å–µ—Ä–µ–¥–Ω—è –∑–∞—Ä–ø–ª–∞—Ç–∞
salary_analysis = data.groupby('profession')['salary'].mean().sort_values()
salary_analysis.plot(kind='bar', figsize=(12,6))
plt.title('–°–µ—Ä–µ–¥–Ω—è –∑–∞—Ä–ø–ª–∞—Ç–∞ –∑–∞ –ø—Ä–æ—Ñ–µ—Å—ñ—è–º–∏')
plt.xlabel('–ü—Ä–æ—Ñ–µ—Å—ñ—è')
plt.ylabel('–°–µ—Ä–µ–¥–Ω—è –∑–∞—Ä–ø–ª–∞—Ç–∞')
plt.show()
```

## –¢–µ–Ω–¥–µ–Ω—Ü—ñ—ó –Ω–∞ —Ä–∏–Ω–∫—É –≤–∞–∫–∞–Ω—Å—ñ–π
–ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–≤–∞–≤—à–∏ –¥–∞–Ω—ñ –∑–∞ –æ—Å—Ç–∞–Ω–Ω—ñ —Ä–æ–∫–∏, –º–∏ –º–æ–∂–µ–º–æ –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –æ—Å–Ω–æ–≤–Ω—ñ —Ç–µ–Ω–¥–µ–Ω—Ü—ñ—ó –Ω–∞ —Ä–∏–Ω–∫—É –≤–∞–∫–∞–Ω—Å—ñ–π –¥–ª—è –∂—ñ–Ω–æ–∫ —É –ü–æ–ª—å—â—ñ. –¶–µ–π –∞–Ω–∞–ª—ñ–∑ –¥–æ–ø–æ–º–æ–∂–µ –∑—Ä–æ–∑—É–º—ñ—Ç–∏, —è–∫—ñ —Å—Ñ–µ—Ä–∏ –∑—Ä–æ—Å—Ç–∞—é—Ç—å —ñ –≤ —è–∫–∏—Ö –Ω–∞–ø—Ä—è–º–∫–∞—Ö –≤–∞—Ä—Ç–æ —à—É–∫–∞—Ç–∏ –Ω–æ–≤—ñ –º–æ–∂–ª–∏–≤–æ—Å—Ç—ñ.
–î–æ–¥–∞—Ç–∫–æ–≤—É —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—é –ø—Ä–æ –º–æ–∂–ª–∏–≤–æ—Å—Ç—ñ –¥–ª—è –∂—ñ–Ω–æ–∫ –≤ –ü–æ–ª—å—â—ñ –º–æ–∂–Ω–∞ –∑–Ω–∞–π—Ç–∏ —É —Å—Ç–∞—Ç—Ç—ñ: [–†–æ–±–æ—Ç–∞ –≤ –ü–æ–ª—å—â—ñ –¥–ª—è –∂—ñ–Ω–æ–∫ –ø–æ –±—ñ–æ–º–µ—Ç—Ä—ñ—ó: –º–æ–∂–ª–∏–≤–æ—Å—Ç—ñ —Ç–∞ –ø–æ—Ä–∞–¥–∏](https://pracaplus.org/article.php?id=82269227-f692-4e26-93a1-5e7de1f5e2c6).

```python
# –ê–Ω–∞–ª—ñ–∑ —Ç–µ–Ω–¥–µ–Ω—Ü—ñ–π
trend_data = data['date_posted'].value_counts().sort_index()
trend_data.plot(kind='line', figsize=(12,6))
plt.title('–¢–µ–Ω–¥–µ–Ω—Ü—ñ—ó –≤–∞–∫–∞–Ω—Å—ñ–π —É —á–∞—Å—ñ')
plt.xlabel('–î–∞—Ç–∞ –ø—É–±–ª—ñ–∫–∞—Ü—ñ—ó –≤–∞–∫–∞–Ω—Å—ñ—ó')
plt.ylabel('–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∞–∫–∞–Ω—Å—ñ–π')
plt.show()
```
