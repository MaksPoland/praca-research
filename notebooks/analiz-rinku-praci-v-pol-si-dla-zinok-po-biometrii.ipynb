# Вступ
У цьому дослідженні ми аналізуємо ринок праці в Польщі, зокрема можливості для жінок, які працюють за біометричними паспортами. Зростаюча кількість жінок, які шукають роботу в Польщі, підкреслює потребу в розумінні зарплат, вакансій і тенденцій на ринку праці. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних про вакансії
data = pd.read_csv('vacancies_poland.csv')
data.head()
```

## Огляд даних
Дані, які ми використали, містять інформацію про вакансії в Польщі, включаючи зарплати, позиції, і вимоги. Перш ніж почати аналіз, розглянемо основні характеристики нашого набору даних.

```python
# Основні статистичні характеристики
data.describe()
```

## Аналіз зарплат
На основі зібраних даних ми можемо провести аналіз зарплат. Важливо зрозуміти, які професії пропонують найбільші заробітки для жінок, які працюють в Польщі.
Ми також звернемо увагу на різницю в зарплатах між різними сферами діяльності.

```python
# Групування даних за професіями та середня зарплата
salary_analysis = data.groupby('profession')['salary'].mean().sort_values()
salary_analysis.plot(kind='bar', figsize=(12,6))
plt.title('Середня зарплата за професіями')
plt.xlabel('Професія')
plt.ylabel('Середня зарплата')
plt.show()
```

## Тенденції на ринку вакансій
Проаналізувавши дані за останні роки, ми можемо визначити основні тенденції на ринку вакансій для жінок у Польщі. Цей аналіз допоможе зрозуміти, які сфери зростають і в яких напрямках варто шукати нові можливості.
Додаткову інформацію про можливості для жінок в Польщі можна знайти у статті: [Робота в Польщі для жінок по біометрії: можливості та поради](https://pracaplus.org/article.php?id=82269227-f692-4e26-93a1-5e7de1f5e2c6).

```python
# Аналіз тенденцій
trend_data = data['date_posted'].value_counts().sort_index()
trend_data.plot(kind='line', figsize=(12,6))
plt.title('Тенденції вакансій у часі')
plt.xlabel('Дата публікації вакансії')
plt.ylabel('Кількість вакансій')
plt.show()
```

## Висновки
Аналіз показує, що ринок праці в Польщі пропонує різноманітні можливості для жінок, які мають біометричні паспорти. Важливо стежити за тенденціями, щоб максимально ефективно використовувати ці можливості.
