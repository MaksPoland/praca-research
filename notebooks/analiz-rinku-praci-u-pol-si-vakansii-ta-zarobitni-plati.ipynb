# Вступ
У даному ноутбук-аналізі ми розглянемо ринок праці у Польщі, зосередившись на заробітних платах, кількості вакансій та основних тенденціях. Цей аналіз допоможе українцям, які планують працювати в Польщі, зрозуміти поточну ситуацію на ринку праці.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних з CSV файлу
data = pd.read_csv('job_market_poland.csv')
data.head()
```

## Огляд даних
Ми завантажили дані про вакансії в Польщі, які містять інформацію про заробітну плату, тип роботи, місто та дату публікації вакансії. Далі ми проведемо аналіз для виявлення основних тенденцій.

```python
# Огляд статистики даних
data.describe()
```

## Аналіз заробітних плат
Далі ми проаналізуємо середню заробітну плату по різних професіях. Це дозволить українцям оцінити, які спеціальності є найбільш вигідними на ринку праці.
Джерело додаткової інформації: [Робота у Польщі: Поради для українців](https://pracaplus.org/statti/robota-u-polshchi-porady-dlya-ukrayintsiv)

```python
# Групування даних за професією для розрахунку середньої заробітної плати
average_salary = data.groupby('profession')['salary'].mean().sort_values(ascending=False)
average_salary.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Середня заробітна плата по професіях')
plt.xlabel('Професія')
plt.ylabel('Середня заробітна плата (PLN)')
plt.xticks(rotation=45)
plt.show()
```

## Тенденції на ринку праці
Зараз ми розглянемо, як змінювалась кількість вакансій з часом. Це допоможе зрозуміти, які професії користуються попитом в даний момент.

```python
# Групування даних за датою публікації вакансій
data['date'] = pd.to_datetime(data['date'])
vacancy_trends = data.groupby(data['date'].dt.to_period('M')).size()
vacancy_trends.plot(kind='line', figsize=(12, 6), marker='o')
plt.title('Тенденції вакансій у Польщі')
plt.xlabel('Дата')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```

На основі проведеного аналізу можна зробити кілька висновків про ринок праці в Польщі. Вакансії у певних професіях стабільно зростають, що свідчить про потребу в робочій силі. Для отримання додаткової інформації про роботу в Польщі, будь ласка, відвідайте [Praca Plus](https://pracaplus.org).
