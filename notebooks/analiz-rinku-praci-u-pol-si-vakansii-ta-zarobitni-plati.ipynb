# –í—Å—Ç—É–ø
–£ –¥–∞–Ω–æ–º—É –Ω–æ—É—Ç–±—É–∫-–∞–Ω–∞–ª—ñ–∑—ñ –º–∏ —Ä–æ–∑–≥–ª—è–Ω–µ–º–æ —Ä–∏–Ω–æ–∫ –ø—Ä–∞—Ü—ñ —É –ü–æ–ª—å—â—ñ, –∑–æ—Å–µ—Ä–µ–¥–∏–≤—à–∏—Å—å –Ω–∞ –∑–∞—Ä–æ–±—ñ—Ç–Ω–∏—Ö –ø–ª–∞—Ç–∞—Ö, –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≤–∞–∫–∞–Ω—Å—ñ–π —Ç–∞ –æ—Å–Ω–æ–≤–Ω–∏—Ö —Ç–µ–Ω–¥–µ–Ω—Ü—ñ—è—Ö. –¶–µ–π –∞–Ω–∞–ª—ñ–∑ –¥–æ–ø–æ–º–æ–∂–µ —É–∫—Ä–∞—ó–Ω—Ü—è–º, —è–∫—ñ –ø–ª–∞–Ω—É—é—Ç—å –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –≤ –ü–æ–ª—å—â—ñ, –∑—Ä–æ–∑—É–º—ñ—Ç–∏ –ø–æ—Ç–æ—á–Ω—É —Å–∏—Ç—É–∞—Ü—ñ—é –Ω–∞ —Ä–∏–Ω–∫—É –ø—Ä–∞—Ü—ñ.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –¥–∞–Ω–∏—Ö –∑ CSV —Ñ–∞–π–ª—É
data = pd.read_csv('job_market_poland.csv')
data.head()
```

## –û–≥–ª—è–¥ –¥–∞–Ω–∏—Ö
–ú–∏ –∑–∞–≤–∞–Ω—Ç–∞–∂–∏–ª–∏ –¥–∞–Ω—ñ –ø—Ä–æ –≤–∞–∫–∞–Ω—Å—ñ—ó –≤ –ü–æ–ª—å—â—ñ, —è–∫—ñ –º—ñ—Å—Ç—è—Ç—å —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—é –ø—Ä–æ –∑–∞—Ä–æ–±—ñ—Ç–Ω—É –ø–ª–∞—Ç—É, —Ç–∏–ø —Ä–æ–±–æ—Ç–∏, –º—ñ—Å—Ç–æ —Ç–∞ –¥–∞—Ç—É –ø—É–±–ª—ñ–∫–∞—Ü—ñ—ó –≤–∞–∫–∞–Ω—Å—ñ—ó. –î–∞–ª—ñ –º–∏ –ø—Ä–æ–≤–µ–¥–µ–º–æ –∞–Ω–∞–ª—ñ–∑ –¥–ª—è –≤–∏—è–≤–ª–µ–Ω–Ω—è –æ—Å–Ω–æ–≤–Ω–∏—Ö —Ç–µ–Ω–¥–µ–Ω—Ü—ñ–π.

```python
# –û–≥–ª—è–¥ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–∞–Ω–∏—Ö
data.describe()
```

## –ê–Ω–∞–ª—ñ–∑ –∑–∞—Ä–æ–±—ñ—Ç–Ω–∏—Ö –ø–ª–∞—Ç
–î–∞–ª—ñ –º–∏ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É—î–º–æ —Å–µ—Ä–µ–¥–Ω—é –∑–∞—Ä–æ–±—ñ—Ç–Ω—É –ø–ª–∞—Ç—É –ø–æ —Ä—ñ–∑–Ω–∏—Ö –ø—Ä–æ—Ñ–µ—Å—ñ—è—Ö. –¶–µ –¥–æ–∑–≤–æ–ª–∏—Ç—å —É–∫—Ä–∞—ó–Ω—Ü—è–º –æ—Ü—ñ–Ω–∏—Ç–∏, —è–∫—ñ —Å–ø–µ—Ü—ñ–∞–ª—å–Ω–æ—Å—Ç—ñ —î –Ω–∞–π–±—ñ–ª—å—à –≤–∏–≥—ñ–¥–Ω–∏–º–∏ –Ω–∞ —Ä–∏–Ω–∫—É –ø—Ä–∞—Ü—ñ.
–î–∂–µ—Ä–µ–ª–æ –¥–æ–¥–∞—Ç–∫–æ–≤–æ—ó —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó: [–†–æ–±–æ—Ç–∞ —É –ü–æ–ª—å—â—ñ: –ü–æ—Ä–∞–¥–∏ –¥–ª—è —É–∫—Ä–∞—ó–Ω—Ü—ñ–≤](https://pracaplus.org/statti/robota-u-polshchi-porady-dlya-ukrayintsiv)

```python
# –ì—Ä—É–ø—É–≤–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö –∑–∞ –ø—Ä–æ—Ñ–µ—Å—ñ—î—é –¥–ª—è —Ä–æ–∑—Ä–∞—Ö—É–Ω–∫—É —Å–µ—Ä–µ–¥–Ω—å–æ—ó –∑–∞—Ä–æ–±—ñ—Ç–Ω–æ—ó –ø–ª–∞—Ç–∏
average_salary = data.groupby('profession')['salary'].mean().sort_values(ascending=False)
average_salary.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('–°–µ—Ä–µ–¥–Ω—è –∑–∞—Ä–æ–±—ñ—Ç–Ω–∞ –ø–ª–∞—Ç–∞ –ø–æ –ø—Ä–æ—Ñ–µ—Å—ñ—è—Ö')
plt.xlabel('–ü—Ä–æ—Ñ–µ—Å—ñ—è')
plt.ylabel('–°–µ—Ä–µ–¥–Ω—è –∑–∞—Ä–æ–±—ñ—Ç–Ω–∞ –ø–ª–∞—Ç–∞ (PLN)')
plt.xticks(rotation=45)
plt.show()
```

## –¢–µ–Ω–¥–µ–Ω—Ü—ñ—ó –Ω–∞ —Ä–∏–Ω–∫—É –ø—Ä–∞—Ü—ñ
–ó–∞—Ä–∞–∑ –º–∏ —Ä–æ–∑–≥–ª—è–Ω–µ–º–æ, —è–∫ –∑–º—ñ–Ω—é–≤–∞–ª–∞—Å—å –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∞–∫–∞–Ω—Å—ñ–π –∑ —á–∞—Å–æ–º. –¶–µ –¥–æ–ø–æ–º–æ–∂–µ –∑—Ä–æ–∑—É–º—ñ—Ç–∏, —è–∫—ñ –ø—Ä–æ—Ñ–µ—Å—ñ—ó –∫–æ—Ä–∏—Å—Ç—É—é—Ç—å—Å—è –ø–æ–ø–∏—Ç–æ–º –≤ –¥–∞–Ω–∏–π –º–æ–º–µ–Ω—Ç.

```python
# –ì—Ä—É–ø—É–≤–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö –∑–∞ –¥–∞—Ç–æ—é –ø—É–±–ª—ñ–∫–∞—Ü—ñ—ó –≤–∞–∫–∞–Ω—Å—ñ–π
data['date'] = pd.to_datetime(data['date'])
vacancy_trends = data.groupby(data['date'].dt.to_period('M')).size()
vacancy_trends.plot(kind='line', figsize=(12, 6), marker='o')
plt.title('–¢–µ–Ω–¥–µ–Ω—Ü—ñ—ó –≤–∞–∫–∞–Ω—Å—ñ–π —É –ü–æ–ª—å—â—ñ')
plt.xlabel('–î–∞—Ç–∞')
plt.ylabel('–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∞–∫–∞–Ω—Å—ñ–π')
plt.grid()
plt.show()
```
