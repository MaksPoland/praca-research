# Вступ
У цьому нотатнику ми проведемо аналіз ринку праці в Ольштині, Польща. Зокрема, ми розглянемо дані про заробітні плати, кількість вакансій та загальні тенденції на ринку праці. Цей аналіз допоможе зрозуміти, які професії користуються попитом, а також які умови працевлаштування в даному регіоні.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('job_market_olsztyn.csv')
data.head()
```

## Огляд даних
Дані, які ми використовуємо, містять інформацію про вакансії в Ольштині, заробітні плати, а також інші важливі характеристики. У таблиці представлені такі стовпці: 'посада', 'зарплата', 'дата публікації' та 'тип зайнятості'.

```python
# Очищення даних
data = data.dropna()  # Видалення пропущених значень
data['дата публікації'] = pd.to_datetime(data['дата публікації'])
data['рік'] = data['дата публікації'].dt.year
data.head()
```

## Аналіз зарплат
Ми проаналізуємо розподіл заробітних плат за різними професіями. Це дозволить виявити, які професії найбільш оплачувані та в яких сферах є попит.

Джерело: [Робота в Ольштині: Перспективи та Умови Працевлаштування](https://pracaplus.org/statti/robota-v-olshtyni-perspektyvy-ta-umovy-pratsevlashtuvannya)

```python
# Візуалізація зарплат по професіях
plt.figure(figsize=(12, 6))
data.groupby('посада')['зарплата'].mean().sort_values().plot(kind='barh')
plt.title('Середня зарплата по професіях в Ольштині')
plt.xlabel('Зарплата (PLN)')
plt.ylabel('Професія')
plt.show()
```

## Тенденції на ринку праці
Дослідимо, як змінювалась кількість вакансій у Ольштині з часом. Це допоможе зрозуміти динаміку ринку праці та виявити можливі сезонні коливання.

```python
# Аналіз тенденцій вакансій
plt.figure(figsize=(12, 6))
vacancies_trend = data.groupby('рік')['посада'].count()
vacancies_trend.plot(kind='line')
plt.title('Кількість вакансій у Ольштині по роках')
plt.xlabel('Рік')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```

## Висновки
У результаті нашого аналізу ми дізналися, що ринок праці в Ольштині демонструє стабільний попит на певні професії. Заробітні плати варіюються в залежності від сфери діяльності, а тенденції показують зростання кількості вакансій в останні роки.
