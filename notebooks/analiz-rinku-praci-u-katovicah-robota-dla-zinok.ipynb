# Вступ
У цьому ноутбуці ми проаналізуємо ринок праці у Катовицях, зокрема можливості працевлаштування для жінок. Згідно з даними, представлення жінок у різних секторах економіки є важливим для розвитку місцевого ринку праці. Ми дослідимо зарплати, вакансії та тенденції на ринку праці для жінок.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних
data = pd.read_csv('job_market_katowice.csv')
data.head()
```

## Огляд даних
Після завантаження даних, ми переглянемо основні характеристики вакансій, доступних для жінок у Катовицях. Це допоможе нам зрозуміти, які професії є найбільш затребуваними.

```python
# Описова статистика
data.describe()
```

## Аналіз зарплат
Зарплати є важливим фактором при виборі роботи. Ми проаналізуємо середні зарплати за різними професіями, які найчастіше пропонуються жінкам у Катовицях. Це дозволить виявити найбільш вигідні пропозиції.

```python
# Аналіз середніх зарплат
average_salaries = data.groupby('profession')['salary'].mean().sort_values(ascending=False)
average_salaries.plot(kind='bar', figsize=(12,6), title='Середня зарплата за професіями')
plt.ylabel('Середня зарплата')
plt.xlabel('Професії')
plt.show()
```

## Тенденції на ринку праці
Регулярний моніторинг вакансій дозволяє виявити тенденції у попиті на спеціальності. Ми проаналізуємо зміни кількості вакансій за останні кілька років.
Цей аналіз містить важливу інформацію для жінок, які шукають роботу у Катовицях. Для додаткових порад можна ознайомитися з статтею [Катовиці: Робота для жінок – можливості та поради](https://pracaplus.org/statti/katovytsi-robota-dlya-zhinok-mozhlyvosti-ta-porady).

```python
# Аналіз вакансій за роками
vacancies_trend = data.groupby('year')['vacancies'].sum()
vacancies_trend.plot(kind='line', figsize=(12,6), marker='o', title='Тенденції вакансій за роками')
plt.ylabel('Кількість вакансій')
plt.xlabel('Рік')
plt.show()
```

## Висновки
Аналіз ринку праці у Катовицях показує, що жінки мають різноманітні можливості для працевлаштування, особливо в таких сферах, як IT, охорона здоров'я та освіта. Важливо стежити за змінами на ринку, щоб бути в курсі нових вакансій і можливостей.
