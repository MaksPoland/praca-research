# Аналіз ринку праці в Познані для жінок
У цьому ноутбуці ми проведемо аналіз ринку праці в Познані, зосередившись на можливостях працевлаштування для жінок. Проаналізуємо дані про зарплати, вакансії та основні тенденції.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Завантаження та підготовка даних
Для аналізу ми скористаємося даними про вакансії в Познані. Ці дані включають інформацію про зарплати, кількість вакансій за різними сферами та інші характеристики.

```python
data = pd.read_csv('job_data_poznan.csv')
data.head()
```

## Огляд вакансій та зарплат
Проаналізуємо дані про зарплати та кількість вакансій для жінок у різних сферах. Цей аналіз допоможе зрозуміти, які професії є найбільш затребуваними.
Джерело: [Робота в Познані для жінок: можливості та поради](https://pracaplus.org/statti/robota-v-poznani-dlya-zhinok-mozhlyvosti-ta-porady)

```python
salary_by_sector = data.groupby('sector')['salary'].mean().sort_values()
salary_by_sector.plot(kind='barh', figsize=(10, 6))
plt.title('Середня зарплата за секторами')
plt.xlabel('Середня зарплата (PLN)')
plt.ylabel('Сектор')
plt.show()
```

## Тенденції на ринку праці
Розглянемо тенденції на ринку праці, зокрема зміну кількості вакансій для жінок за останні кілька років. Це дозволить зрозуміти, чи зростає попит на жіночу працю в Познані.

```python
vacancies_trend = data.groupby('year')['vacancies'].sum()
vacancies_trend.plot(kind='line', figsize=(10, 6))
plt.title('Тенденція кількості вакансій для жінок')
plt.xlabel('Рік')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```
