# Вступ
У цьому ноутбуці ми проведемо аналіз ринку праці в Бидгощі для українців, які шукають роботу в Польщі. Проаналізуємо дані про зарплати, вакансії та загальні тенденції на ринку праці.

Джерело: [Бидгощ: робота для українців у Польщі](https://pracaplus.org/statti/bydhoshch-robota-dlya-ukrayintsiv-u-polshchi)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних про вакансії
data = pd.read_csv('bydgoszcz_jobs.csv')
data.head()
```

# Огляд даних
У таблиці ми маємо інформацію про вакансії, їх зарплати, вимоги до кандидатів та інші деталі. Далі ми проведемо деякі базові аналізи даних.

```python
# Аналіз середньої зарплати по вакансіях
average_salary = data['salary'].mean()
print(f'Середня зарплата у Бидгощі: {average_salary:.2f} PLN')
```

# Тенденції на ринку праці
Аналізуючи дані, можна побачити, що попит на певні професії зріс, особливо в IT-секторі. Це може бути корисним для українців, які планують працювати в цій сфері.

```python
# Візуалізація розподілу зарплат
plt.figure(figsize=(10, 6))
plt.hist(data['salary'], bins=20, color='blue', alpha=0.7) 
plt.title('Розподіл зарплат у Бидгощі')
plt.xlabel('Зарплата (PLN)')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```
