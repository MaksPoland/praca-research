# Вступ
У цьому дослідженні ми проаналізуємо ринок праці в Карпачі, Польща, зокрема можливості для українців. Карпач є популярним курортом, що пропонує різноманітні вакансії у сфері туризму, hospitality та обслуговування. Ми розглянемо статистику заробітних плат, кількість вакансій та основні тенденції.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних про вакансії та заробітні плати
data = pd.read_csv('job_market_karpach.csv')
data.head()
```

## Огляд даних
Ми завантажили дані про вакансії та заробітні плати в Карпачі. Далі ми проведемо огляд та очистку даних, щоб підготувати їх для аналізу.
Дані включають такі колонки, як: 'посада', 'зарплата', 'тип зайнятості', 'дата публікації'.

```python
# Очищення даних
data.dropna(inplace=True)
data['зарплата'] = data['зарплата'].str.replace('€', '').str.replace('.', '').astype(float)
data['дата публікації'] = pd.to_datetime(data['дата публікації'])
data.describe()
```

## Аналіз заробітних плат
Ми провели аналіз заробітних плат в Карпачі. На основі очищених даних ми можемо візуалізувати розподіл заробітних плат за різними посадовими категоріями.
Зверніть увагу, що середня заробітна плата для вакансій у сфері hospitality вища за середню по ринку.

```python
# Візуалізація розподілу заробітних плат
plt.figure(figsize=(10, 6))
data['зарплата'].hist(bins=30, color='skyblue')
plt.title('Розподіл заробітних плат у Карпачі')
plt.xlabel('Зарплата (€)')
plt.ylabel('Кількість вакансій')
plt.grid(False)
plt.show()
```

## Висновки
Аналіз показує, що ринок праці в Карпачі пропонує різноманітні можливості для українців. Середня заробітна плата у сфері обслуговування та туризму є конкурентоспроможною.
