```python
average_salaries = data.groupby('position')['salary'].mean().sort_values()
average_salaries.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('–°–µ—Ä–µ–¥–Ω—ñ –ó–∞—Ä–æ–±—ñ—Ç–Ω—ñ –ü–ª–∞—Ç–∏ –ø–æ –í–∞–∫–∞–Ω—Å—ñ—è—Ö —É –ö–∞—Ç–æ–≤—ñ—Ü–µ')
plt.xlabel('–°–µ—Ä–µ–¥–Ω—è –ó–∞—Ä–æ–±—ñ—Ç–Ω–∞ –ü–ª–∞—Ç–∞ (PLN)')
plt.ylabel('–í–∞–∫–∞–Ω—Å—ñ—ó')
plt.show()
```
