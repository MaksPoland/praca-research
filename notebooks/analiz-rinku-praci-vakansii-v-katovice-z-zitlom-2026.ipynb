# Вступ
У цьому ноутбуці ми проведемо аналіз вакансій у Катовіце, Польща, на 2026 рік. Зокрема, ми зосередимося на вакансіях, які пропонують житло, оскільки це є важливим фактором для багатьох шукачів роботи. Аналіз включатиме огляд заробітних плат, тенденцій на ринку праці та кількості доступних вакансій.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних вакансій
data = pd.read_csv('vacancies_katowice_2026.csv')
data.head()
```

## Огляд Даних
Ми завантажили дані про вакансії в Катовіце з житлом на 2026 рік. Давайте проаналізуємо структуру даних та виведемо основні характеристики набору даних.

```python
data.info()
data.describe()
```

## Аналіз Заробітних Плат
У цьому розділі ми розглянемо середні заробітні плати по різним вакансіям. Це допоможе зрозуміти, які професії є найбільш прибутковими у Катовіце.

```python
average_salaries = data.groupby('position')['salary'].mean().sort_values()
average_salaries.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Середні Заробітні Плати по Вакансіях у Катовіце')
plt.xlabel('Середня Заробітна Плата (PLN)')
plt.ylabel('Вакансії')
plt.show()
```

## Тенденції на Ринку Праці
Давайте розглянемо, як змінювалися вакансії з житлом протягом останніх років. Це дозволить нам зрозуміти загальні тенденції на ринку праці у Катовіце.
Джерело інформації: [Вакансії в Катовіце з Житлом 2026](https://pracaplus.org/article.php?id=d9b8d262-812d-4761-a864-50d38ffe1cf3)

```python
trend_data = pd.read_csv('vacancies_trend_katowice.csv')
plt.plot(trend_data['year'], trend_data['vacancies'], marker='o')
plt.title('Тенденції Вакансій у Катовіце з Житлом')
plt.xlabel('Рік')
plt.ylabel('Кількість Вакансій')
plt.grid()
plt.show()
```
