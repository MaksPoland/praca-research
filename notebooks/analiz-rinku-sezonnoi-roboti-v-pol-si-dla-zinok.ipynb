# Вступ
Цей блокнот присвячений аналізу ринку сезонної роботи в Польщі для жінок. Зокрема, ми розглянемо дані про вакансії, зарплати та тенденції на ринку праці. Сезонна робота є важливим джерелом доходу для багатьох жінок, які шукають можливості працевлаштування за кордоном.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('seasonal_jobs_poland.csv')
data.head()
```

# Огляд даних
Дані містять інформацію про вакансії, включаючи заробітну плату, галузь, місто та інші характеристики. Ми проаналізуємо, які вакансії є найбільш популярними серед жінок та які пропонують найвищі зарплати.

```python
# Фільтрація даних за категорією 'жінки'
women_jobs = data[data['gender'] == 'female']
women_jobs.describe()
```

# Аналіз зарплат
Аналізуючи зарплати для жінок, ми можемо виявити, які галузі пропонують найбільші можливості заробітку. Це дозволяє жінкам краще планувати свій кар'єрний шлях.
Додаткову інформацію про можливості сезонної роботи в Польщі можна знайти в статті [Сезонна робота в Польщі для жінок: можливості та поради](https://pracaplus.org/statti/sezonna-robota-v-polshchi-dlya-zhinok-mozhlyvosti-ta-porady).

```python
# Візуалізація зарплат за галузями
plt.figure(figsize=(10, 6))
women_jobs.groupby('industry')['salary'].mean().sort_values().plot(kind='bar')
plt.title('Середня зарплата жінок за галузями')
plt.xlabel('Галузь')
plt.ylabel('Середня зарплата')
plt.xticks(rotation=45)
plt.show()
```

# Тенденції на ринку праці
Завдяки аналізу даних, ми можемо виявити тенденції на ринку праці. Це дозволяє жінкам краще орієнтуватися у виборі вакансій, що відповідають їхнім здібностям та інтересам.
Відвідайте основний сайт для отримання додаткової інформації: [Praca Plus](https://pracaplus.org).

```python
# Аналіз тенденцій за роками
trend_data = data.groupby('year')['vacancies'].count()
plt.figure(figsize=(10, 6))
trend_data.plot(kind='line')
plt.title('Тенденції вакансій за роками')
plt.xlabel('Рік')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```

# Висновки
Аналіз ринку сезонної роботи в Польщі для жінок показує, що існує багато можливостей для працевлаштування. Однак, важливо звертати увагу на зарплати та тенденції, щоб зробити оптимальний вибір.
