# Вступ
У цьому нотатнику ми проаналізуємо ринок праці в Польщі для осіб, які мають біометричні паспорти. Ми розглянемо наявні вакансії, середні заробітні плати, а також основні тенденції на ринку праці.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних про вакансії
data = pd.read_csv('vacancies_poland.csv')
data.head()
```

## Огляд даних
Ми отримали дані про вакансії в Польщі, які містять інформацію про професію, заробітну плату, місцезнаходження та інші важливі параметри. У наступних розділах ми розглянемо ці дані більш детально.

```python
# Огляд загальних характеристик даних
data.describe()
```

## Аналіз заробітних плат
У цьому розділі ми проаналізуємо середні заробітні плати по різних професіях, щоб зрозуміти, які вакансії є найбільш вигідними.
Джерело: [Робота в Польщі по біометричному паспорту: вакансії та можливості](https://pracaplus.org/statti/robota-v-polshchi-po-biometrychnomu-pasportu-vakansiyi-ta-mozhlyvosti)

```python
# Групування даних за професією та обчислення середньої заробітної плати
salary_analysis = data.groupby('profession')['salary'].mean().sort_values(ascending=False)
salary_analysis.plot(kind='bar', figsize=(12, 6), title='Середня заробітна плата за професією')
plt.ylabel('Заробітна плата (PLN)')
plt.xlabel('Професія')
plt.show()
```

## Тенденції на ринку праці
У цьому розділі ми розглянемо загальні тенденції на ринку вакансій в Польщі для власників біометричних паспортів. Це допоможе нам зрозуміти, як змінюється попит на різні професії.

```python
# Аналіз тенденцій за часом
trends = data['date_posted'].value_counts().sort_index()
trends.plot(figsize=(12, 6), title='Тенденції розміщення вакансій з часом')
plt.ylabel('Кількість вакансій')
plt.xlabel('Дата')
plt.show()
```

## Висновки
Аналіз показав, що ринок праці в Польщі пропонує широкий спектр вакансій для осіб з біометричними паспортами. Середні заробітні плати варіюються, і деякі професії є особливо вигідними. Для отримання додаткової інформації, відвідайте [Praca Plus](https://pracaplus.org).
