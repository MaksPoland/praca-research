# Вступ
У цьому ноутбуці ми проведемо аналіз ринку праці в Польщі, зосередившись на вакансіях, пов'язаних зі збором клубники. Відзначимо поточні тенденції, зарплати та прогнози для 2026 року.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних про вакансії
data = pd.read_csv('job_market_poland_strawberry.csv')
data.head()
```

## Огляд даних
Ми отримали дані про вакансії, які включають такі стовпці, як назва позиції, зарплата, регіон, і кількість вакансій. Це дозволить нам зрозуміти основні тенденції на ринку праці.

```python
# Оцінка зарплат
salary_stats = data['salary'].describe()
salary_stats
```

### Статистика зарплат
Згідно з отриманими даними, середня зарплата для вакансій, пов'язаних зі збором клубники, становить близько 3000 PLN на місяць. Це свідчить про значний попит на дані позиції в сезон збору.

```python
# Візуалізація розподілу зарплат
plt.figure(figsize=(10, 6))
plt.hist(data['salary'], bins=20, color='skyblue', edgecolor='black')
plt.title('Розподіл зарплат для вакансій з збору клубники')
plt.xlabel('Зарплата (PLN)')
plt.ylabel('Кількість вакансій')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Тенденції на ринку праці
Як ми можемо побачити на графіку, більшість вакансій пропонують зарплати в межах 2500-3500 PLN. Ці дані можуть бути корисними для тих, хто планує працювати в Польщі в 2026 році. Для детальнішої інформації про прогнози на 2026 рік, ви можете ознайомитися з статтею [Робота в Польщі: Збір клубники в 2026 році](https://pracaplus.org/statti/robota-v-polshchi-zbir-klubnyky-v-2026-rotsi).

```python
# Аналіз вакансій за регіонами
region_counts = data['region'].value_counts()
region_counts.plot(kind='bar', figsize=(12, 6), color='lightgreen')
plt.title('Кількість вакансій за регіонами')
plt.xlabel('Регіон')
plt.ylabel('Кількість вакансій')
plt.xticks(rotation=45)
plt.show()
```
