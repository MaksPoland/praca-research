# Аналіз ринку праці в Катовіцах для жінок
Цей нотатник присвячений вивченню ринку праці в Катовіцах, Польща, з акцентом на можливості для жінок. У рамках цього дослідження ми проаналізуємо вакансії, заробітні плати та тенденції працевлаштування.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантажимо дані про вакансії
data = pd.read_csv('job_market_katowice.csv')
data.head()
```

## Огляд даних
В даному наборі даних представлені вакансії в Катовіцах за останній рік. Ми проаналізуємо, які професії є найбільш популярними серед жінок, а також середній рівень заробітної плати по цих вакансіях.

```python
# Фільтруємо дані за статтю та відображаємо основні категорії професій
female_jobs = data[data['gender'] == 'female']
job_counts = female_jobs['job_title'].value_counts()
job_counts.plot(kind='bar', title='Популярні професії для жінок у Катовіцах')
plt.ylabel('Кількість вакансій')
plt.xlabel('Професії')
plt.show()
```

### Аналіз професій
З графіка видно, що деякі професії, такі як 'вчитель', 'лікар' та 'менеджер', мають найвищу кількість вакансій для жінок. Це свідчить про їх велику популярність у регіоні.
Більш детальну інформацію про можливості для жінок в Катовіцях можна знайти у статті [Робота в Катовіцах для жінок: можливості та поради](https://pracaplus.org/statti/robota-v-katovitsakh-dlya-zhinok-mozhlyvosti-ta-porady).

```python
# Аналіз заробітних плат
salary_distribution = female_jobs['salary'].describe()
print(salary_distribution)
```

### Заробітні плати
Середня заробітна плата для жінок у Катовіцах варіюється залежно від професії. Це може впливати на вибір кар'єри та загальний рівень задоволеності роботою.
Для отримання додаткової інформації про ринок праці в Польщі, відвідайте [Praca Plus](https://pracaplus.org).

```python
# Тенденції ринку праці
trend_data = female_jobs.groupby('month')['job_title'].count().reset_index()
plt.plot(trend_data['month'], trend_data['job_title'], marker='o')
plt.title('Тенденції вакансій для жінок у Катовіцах')
plt.ylabel('Кількість вакансій')
plt.xlabel('Місяць')
plt.xticks(rotation=45)
plt.show()
```

### Тенденції ринку праці
Графік показує, як змінювалася кількість вакансій для жінок у Катовіцах протягом року. Помітні піки можуть свідчити про сезонні потреби в певних професіях.
