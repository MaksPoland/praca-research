# Вступ
У цьому ноутбуці ми проаналізуємо ринок праці в Варшаві, зосередившись на вакансіях у секторі будівництва. Ми розглянемо дані про заробітні плати, кількість вакансій та основні тенденції, що спостерігаються в цій галузі.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('job_market_warsaw_construction.csv')
data.head()
```

## Огляд даних
Ми завантажили дані про вакансії будівельної галузі у Варшаві. Далі ми детально проаналізуємо отримані дані, щоб зрозуміти основні тенденції та середні заробітні плати.

```python
# Очищення даних
data.dropna(inplace=True)
data['salary'] = data['salary'].str.replace('PLN', '').str.replace(',', '').astype(float)
data['vacancies'] = data['vacancies'].str.replace(',', '').astype(int)
data.describe()
```

## Аналіз заробітних плат
На графіку нижче ми можемо побачити розподіл заробітних плат для вакансій у будівництві в Варшаві. Це дозволяє зрозуміти, які суми пропонуються роботодавцями.

```python
plt.figure(figsize=(10, 6))
plt.hist(data['salary'], bins=30, color='blue', alpha=0.7)
plt.title('Розподіл заробітних плат у будівництві Варшави')
plt.xlabel('Заробітна плата (PLN)')
plt.ylabel('Кількість вакансій')
plt.grid(axis='y')
plt.show()
```

## Тенденції на ринку праці
Згідно з даними, на ринку праці у Варшаві в секторі будівництва спостерігається зростання кількості вакансій. Це пов'язано з активним розвитком інфраструктури та великими будівельними проектами.
Додаткову інформацію про вакансії та поради щодо працевлаштування можна знайти в статті [Робота у Варшаві на будові: як знайти вакансії та уникнути підводних каменів](https://pracaplus.org/statti/robota-u-varshavi-na-budovi-yak-znayty-vakansiyi-ta-unyknuty-pidvodnykh-kameniv).

```python
# Аналіз кількості вакансій по місяцях
monthly_vacancies = data.groupby('month')['vacancies'].sum()
plt.figure(figsize=(10, 6))
monthly_vacancies.plot(kind='line', marker='o', color='orange')
plt.title('Кількість вакансій у будівництві по місяцях')
plt.xlabel('Місяць')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```
