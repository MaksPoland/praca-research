# Вступ
У цьому нотатнику ми проведемо аналіз ринку праці в Польщі, зокрема теми "робота в Польщі у пана". Ми розглянемо дані про заробітні плати, вакансії та тенденції на ринку праці, щоб зрозуміти, які можливості працевлаштування доступні для українців в Польщі.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантажуємо дані про вакансії
data = pd.read_csv('job_market_poland.csv')
data.head()
```

## Огляд даних
У таблиці вище представлені дані про вакансії в Польщі, включаючи інформацію про заробітні плати, місце роботи та вимоги до кандидатів. Аналізуючи ці дані, ми зможемо виявити популярні сфери діяльності та середні заробітні плати в різних галузях.

```python
# Аналіз середніх заробітних плат
average_salaries = data.groupby('industry')['salary'].mean().sort_values(ascending=False)
average_salaries.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Середні заробітні плати за галузями')
plt.xlabel('Галузь')
plt.ylabel('Середня зарплата (PLN)')
plt.xticks(rotation=45)
plt.show()
```

## Аналіз заробітних плат
На графіку вище ми бачимо середні заробітні плати за різними галузями. Це дає нам уявлення про те, в яких сферах працюють українці та які заробітки вони можуть очікувати. Наприклад, IT-сфера показує найбільші заробітки, тоді як у деяких традиційних галузях зарплата є нижчою.

```python
# Аналіз кількості вакансій за місяцями
data['date'] = pd.to_datetime(data['date'])
monthly_vacancies = data.resample('M', on='date').size()
monthly_vacancies.plot(figsize=(10, 6), marker='o', color='orange')
plt.title('Кількість вакансій за місяцями')
plt.xlabel('Місяць')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```

## Тенденції на ринку праці
Графік вище показує кількість вакансій за місяцями. Ми можемо помітити, що ринок праці має сезонні коливання, з піками у деякі місяці.
