# Аналіз ринку праці в Катовіцях
У цьому ноутбуці ми проведемо аналіз ринку праці в Катовіцях, зосереджуючи увагу на вакансіях та зарплатах, описаних на платформі OLX. Дослідження базується на даних про вакансії, які допоможуть зрозуміти сучасні тенденції на ринку праці.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних з CSV файлу
data = pd.read_csv('katowice_job_data.csv')
data.head()
```

## Огляд даних
У цьому розділі ми ознайомимося з форматом та основними характеристиками даних. Дані містять інформацію про різні вакансії в Катовіцях, включаючи назву посади, зарплату, тип зайнятості та дату публікації.

```python
# Опис даних
data.describe()
```

## Аналіз зарплат
Визначимо середню зарплату для різних вакансій. Це дозволить побачити, які позиції найбільше оплачуються, а також виявити тенденції на ринку праці. Зверніть увагу, що дані взято з джерела [Робота в Катовіце: огляд вакансій на OLX](https://pracaplus.org/statti/robota-v-katovitse-ohlyad-vakansiy-na-olx).

```python
# Візуалізація середніх зарплат
average_salaries = data.groupby('position')['salary'].mean().sort_values(ascending=False)
average_salaries.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Середні зарплати за вакансіями в Катовіцях')
plt.xlabel('Вакансії')
plt.ylabel('Середня зарплата (PLN)')
plt.xticks(rotation=45)
plt.show()
```

## Тенденції на ринку праці
Останніми роками ринок праці в Катовіцях демонструє позитивну динаміку. Зростання кількості вакансій у різних сферах, зокрема в IT та виробництві, свідчить про економічний розвиток регіону. Для отримання більше інформації про вакансії, відвідайте сайт [Praca Plus](https://pracaplus.org).

```python
# Аналіз кількості вакансій за роками
data['date'] = pd.to_datetime(data['date'])
vacancies_per_year = data['date'].dt.year.value_counts().sort_index()
vacancies_per_year.plot(kind='line', marker='o', figsize=(12, 6), color='orange')
plt.title('Кількість вакансій за роками в Катовіцях')
plt.xlabel('Рік')
plt.ylabel('Кількість вакансій')
plt.grid(True)
plt.show()
```
