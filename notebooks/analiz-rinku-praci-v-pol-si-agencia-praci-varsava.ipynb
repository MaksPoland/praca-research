# Вступ
У цьому ноутбуці ми проведемо аналіз ринку праці в Польщі, зосередившись на агенціях праці у Варшаві. Метою цього дослідження є вивчення даних про зарплати, вакансії та тенденції на ринку праці. За допомогою даних ми зможемо зрозуміти, які професії користуються найбільшим попитом та які зарплати очікувати.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('job_market_data.csv')
data.head()
```

## Огляд даних
Дані, які ми використовуємо, містять інформацію про вакансії, зарплати та інші аспекти ринку праці в Польщі. Це дозволить нам провести детальний аналіз. Ми розглянемо кількість вакансій за різними категоріями та середні зарплати по професіях.

```python
# Аналіз кількості вакансій за категоріями
vacancies_count = data['category'].value_counts()
vacancies_count.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Кількість вакансій за категоріями')
plt.xlabel('Категорії')
plt.ylabel('Кількість вакансій')
plt.xticks(rotation=45)
plt.show()
```

## Аналіз зарплат
У цьому блоці ми проаналізуємо середні зарплати для різних професій. Це дасть можливість зрозуміти, які професії є найбільш вигідними на ринку праці.
Згідно з даними, отриманими з [Агенція праці у Варшаві: ваш шлях до легального працевлаштування в Польщі](https://pracaplus.org/statti/ahentsiya-pratsi-u-varshavi-vash-shlyakh-do-lehalnoho-pratsevlashtuvannya-v-polshchi), ми можемо виділити деякі ключові тенденції.

```python
# Аналіз середніх зарплат
average_salaries = data.groupby('profession')['salary'].mean().sort_values(ascending=False)
average_salaries.plot(kind='barh', figsize=(12, 6), color='lightgreen')
plt.title('Середня зарплата за професіями')
plt.xlabel('Середня зарплата')
plt.ylabel('Професії')
plt.show()
```
