# Вступ
У цьому дослідженні ми проведемо аналіз ринку праці в Польщі для професії карщик. Зокрема, ми розглянемо дані про зарплати, кількість вакансій та актуальні тренди на ринку. Це допоможе зрозуміти, які можливості існують для тих, хто бажає працювати в цій сфері.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних про вакансії карщиків в Польщі
data = pd.read_csv('job_data_poland_forklifts.csv')
data.head()
```

## Огляд даних
Після завантаження даних, ми можемо переглянути перші кілька рядків, щоб зрозуміти структуру нашого набору даних. Цей набір містить інформацію про вакансії, включаючи зарплати, геолокацію та вимоги до кандидатів.

```python
# Очищення даних
data.dropna(inplace=True)
data['salary'] = data['salary'].str.replace('₴', '').astype(float)
data['salary'] = data['salary'].replace(',', '.', regex=True).astype(float)
data.describe()
```

## Аналіз зарплат
Після очищення даних, ми можемо проаналізувати середні зарплати карщиків в різних регіонах Польщі. Це надасть уявлення про те, де є найбільші можливості для працевлаштування.
Також можемо помітити, що зарплата карщиків варіюється залежно від досвіду та специфікацій роботи.

```python
plt.figure(figsize=(12, 6))
data.groupby('region')['salary'].mean().sort_values().plot(kind='barh')
plt.title('Середня зарплата карщиків в різних регіонах Польщі')
plt.xlabel('Середня зарплата (грн)')
plt.ylabel('Регіон')
plt.show()
```

## Тренди на ринку праці
На основі даних, ми можемо також виявити тренди зростання чи зменшення кількості вакансій у певних регіонах. Це важливо для оцінки перспективності працевлаштування.
Згідно з дослідженням, яке можна знайти на сайті [Praca Plus](https://pracaplus.org), ринок праці для карщиків в Польщі стабільно зростає.

```python
# Графік кількості вакансій по роках
vacancies_over_time = data.groupby('year')['vacancies'].sum()
plt.figure(figsize=(12, 6))
vacancies_over_time.plot(kind='line')
plt.title('Кількість вакансій карщиків в Польщі по роках')
plt.xlabel('Рік')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```
