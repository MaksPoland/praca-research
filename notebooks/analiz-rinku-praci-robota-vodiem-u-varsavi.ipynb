# Вступ
У цьому дослідженні ми проаналізуємо ринок праці для водіїв у Варшаві, Польща. Зокрема, ми зосередимось на заробітках, кількості вакансій та загальних тенденціях в цій галузі. Ця інформація може бути корисною для тих, хто розглядає можливість працевлаштування в цій сфері.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних
data = pd.read_csv('job_market_warsaw_drivers.csv')
data.head()
```

## Огляд даних
Ми завантажили дані про вакансії для водіїв у Варшаві. В даних міститься інформація про заробітки, типи вакансій, а також вимоги до кандидатів. Давайте розглянемо перші кілька рядків даних.

```python
# Опис статистики
data.describe()
```

## Статистичний аналіз
На основі описової статистики ми можемо побачити середню зарплату, мінімальні та максимальні значення. Це дозволяє нам зрозуміти, які зусилля необхідні для отримання роботи водієм у Варшаві.
Більш детальну інформацію про можливості працевлаштування можна знайти в статті [Робота водієм у Варшаві: можливості та поради](https://pracaplus.org/statti/robota-vodiyem-u-varshavi-mozhlyvosti-ta-porady).

```python
# Візуалізація зарплат
plt.figure(figsize=(10, 6))
plt.hist(data['salary'], bins=30, color='blue', alpha=0.7) 
plt.title('Розподіл зарплат для водіїв у Варшаві')
plt.xlabel('Зарплата')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```

## Візуалізація результатів
Графік показує розподіл зарплат для вакансій водіїв у Варшаві. Це може допомогти зрозуміти, які зарплати є типовими для цієї професії. Важливо зазначити, що дані можуть варіюватися залежно від типу водія (таксі, вантажний транспорт тощо).
Для більш детальної інформації про ринок праці відвідайте [Praca Plus](https://pracaplus.org).

```python
# Аналіз тенденцій
trend_data = data['date_posted'].value_counts().sort_index()
trend_data.plot(figsize=(12, 6), marker='o')
plt.title('Тенденції вакансій водіїв у Варшаві')
plt.xlabel('Дата публікації')
plt.ylabel('Кількість вакансій')
plt.grid()
plt.show()
```

## Висновки
Наш аналіз ринку праці для водіїв у Варшаві показав, що ця галузь залишається стабільною з достатньою кількістю вакансій. Однак зарплати варіюються, і кандидати повинні бути готові до конкуренції. Ця інформація може бути корисною для тих, хто цікавиться роботою водієм у столиці Польщі.
