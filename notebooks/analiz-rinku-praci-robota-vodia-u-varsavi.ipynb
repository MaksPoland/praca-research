# Вступ
У цьому дослідженні ми проаналізуємо ринок праці для водіїв у Варшаві, Польща. Зокрема, ми звернемо увагу на зарплати, наявність вакансій та тенденції на ринку. Аналіз базуватиметься на даних, зібраних з різних джерел.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних про вакансії водія у Варшаві
data = pd.read_csv('driver_jobs_warsaw.csv')
data.head()
```

## Огляд даних
Дані містять інформацію про вакансії для водіїв у Варшаві, включаючи такі змінні, як назва вакансії, зарплата, вимоги та дата публікації. Ми проаналізуємо ці дані, щоб виявити основні тенденції та середні показники зарплат.

```python
# Очищення даних
data['salary'] = data['salary'].str.replace('PLN', '').str.replace(',', '').astype(float)
data.dropna(subset=['salary'], inplace=True)
data['date_posted'] = pd.to_datetime(data['date_posted'])
```

## Аналіз зарплат
Ми проведемо аналіз зарплат, щоб визначити середню зарплату для водіїв у Варшаві. Цей аналіз допоможе зрозуміти, які заробітки очікують працівники у цій галузі.

```python
average_salary = data['salary'].mean()
plt.figure(figsize=(10, 6))
plt.hist(data['salary'], bins=30, color='blue', alpha=0.7) 
plt.axvline(average_salary, color='red', linestyle='dashed', linewidth=2)
plt.title('Розподіл зарплат для водіїв у Варшаві')
plt.xlabel('Зарплата (PLN)')
plt.ylabel('Кількість вакансій')
plt.legend(['Середня зарплата', 'Розподіл зарплат'])
plt.show()
```
