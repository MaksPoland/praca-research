# Вступ
У цьому дослідженні ми проаналізуємо ринок праці в Кемпно, Польща, зосередившись на зарплатах, вакансиях та тенденціях. Кемпно, як мале місто, має свої особливості, які впливають на можливості працевлаштування для українців.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних про вакансії
data = pd.read_csv('data/vacancies_kempno.csv')
data.head()
```

## Огляд даних
Ми завантажили дані про вакансії в Кемпно, які містять інформацію про позиції, зарплати та вимоги. Перш ніж продовжувати, розглянемо основні показники.

```python
# Описові статистики для зарплат
salary_stats = data['salary'].describe()
salary_stats
```

## Аналіз зарплат
Згідно з отриманими даними, середня зарплата у Кемпно складає приблизно X PLN. Це показує, що ринок праці може бути вигідним для кандидатів з відповідними кваліфікаціями. Більш детальну інформацію можна знайти у статті [Робота в Польщі: Кемпно – можливості для українців](https://pracaplus.org/statti/robota-v-polshchi-kempno-mozhlyvosti-dlya-ukrayintsiv).

```python
# Візуалізація зарплат
plt.figure(figsize=(10, 6))
plt.hist(data['salary'], bins=20, color='blue', alpha=0.7) 
plt.title('Розподіл зарплат у Кемпно')
plt.xlabel('Зарплата (PLN)')
plt.ylabel('Кількість вакансій')
plt.grid(axis='y')
plt.show()
```
