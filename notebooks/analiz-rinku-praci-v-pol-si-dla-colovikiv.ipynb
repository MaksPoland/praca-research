# Аналіз ринку праці в Польщі для чоловіків
Цей дослідницький проект присвячений аналізу ринку праці в Польщі, зокрема вакансій та зарплат для чоловіків. Метою є вивчення тенденцій, можливостей та викликів, з якими стикаються чоловіки, які шукають роботу в Польщі.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних вакансій
data = pd.read_csv('job_data_poland.csv')
data.head()
```

## Огляд даних
У таблиці представлено дані про вакансії, зокрема посади, зарплати та регіони. Далі ми проаналізуємо, які саме вакансії користуються найбільшим попитом серед чоловіків.

```python
# Фільтрація даних для вакансій, що цікавлять чоловіків
male_jobs = data[data['gender_preference'] == 'male']
job_counts = male_jobs['job_title'].value_counts()
job_counts.head(10).plot(kind='bar')
plt.title('Топ-10 вакансій для чоловіків у Польщі')
plt.xlabel('Посада')
plt.ylabel('Кількість вакансій')
plt.show()
```

## Тенденції на ринку праці
За даними, які ми отримали, можна зробити висновок, що найбільший попит мають вакансії у сферах будівництва, IT та логістики. Це стосується як зарплат, так і кількості відкритих вакансій.
Детальніше про можливості для чоловіків в Польщі можна знайти в статті [Робота в Польщі для чоловіків: можливості та рекомендації](https://pracaplus.org/statti/robota-v-polshchi-dlya-cholovikiv-mozhlyvosti-ta-rekomendatsiyi).

```python
# Аналіз зарплат
salary_distribution = male_jobs['salary']
plt.hist(salary_distribution, bins=30, color='blue', alpha=0.7) 
plt.title('Розподіл зарплат для чоловіків у Польщі')
plt.xlabel('Зарплата')
plt.ylabel('Кількість вакансій')
plt.show()
```

## Висновки
Аналіз показав, що середня зарплата для чоловіків у найбільш популярних вакансіях варіюється. Також існує тенденція до зростання зарплат у деяких секторах. Для детальної інформації про ринок праці в Польщі, ви можете відвідати [Praca Plus](https://pracaplus.org).
