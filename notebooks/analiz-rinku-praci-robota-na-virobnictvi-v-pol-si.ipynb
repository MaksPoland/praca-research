У цьому нотатнику ми проведемо аналіз ринку праці в Польщі, зосереджуючи увагу на вакансіях та зарплатах у секторі виробництва. Виробничий сектор є важливою частиною економіки Польщі, і розуміння його динаміки є критично важливим для тих, хто шукає роботу в цій галузі.

## Дослідження даних
Ми використовуватимемо дані про вакансії, зарплати та тренди на ринку праці. Такі дані можуть допомогти зрозуміти, які професії є найбільш затребуваними та які зарплати пропонуються кандидатам.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('job_market_data.csv')
data.head()
```

## Попередній аналіз даних
Після завантаження даних, ми проведемо попередній аналіз, щоб оцінити їх структуру та виявити можливі аномалії. Це допоможе нам краще зрозуміти, як працюють наші дані.

```python
data.info()
data.describe()
```

## Аналіз зарплат у виробництві
Зараз ми зосередимося на аналізі зарплат у виробничому секторі. Розглянемо середні зарплати за різними професіями та змінимо їх у часі.
Згідно з інформацією з джерела [Робота на виробництві в Польщі: повний гід](https://pracaplus.org/article.php?id=test-new-platforms), важливо знати, які зарплати є конкурентоспроможними.

```python
average_salaries = data.groupby('profession')['salary'].mean()
average_salaries.plot(kind='bar', figsize=(10, 6), title='Середні зарплати за професіями у виробництві')
plt.xlabel('Професія')
plt.ylabel('Середня зарплата (PLN)')
plt.show()
```

## Тренди на ринку праці
Наступним кроком буде огляд трендів на ринку праці. Ми проаналізуємо, як змінювались вакансії в виробничому секторі протягом останніх років.

```python
vacancy_trends = data['year'].value_counts().sort_index()
vacancy_trends.plot(kind='line', figsize=(10, 6), title='Тренди вакансій у виробництві')
plt.xlabel('Рік')
plt.ylabel('Кількість вакансій')
plt.show()
```
