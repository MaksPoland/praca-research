## Аналіз зарплат
У цьому розділі ми візуалізуємо розподіл заробітних плат для різних професій. Це дозволить зрозуміти, які професії є найбільш вигідними.
![Різні професії та зарплати](https://example.com/salary_distribution.png)

```python
# Візуалізація зарплат
plt.figure(figsize=(10, 6))
data['salary'].hist(bins=30, color='blue', edgecolor='black')
plt.title('Розподіл заробітних плат')
plt.xlabel('Зарплата (PLN)')
plt.ylabel('Кількість вакансій')
plt.show()
```
