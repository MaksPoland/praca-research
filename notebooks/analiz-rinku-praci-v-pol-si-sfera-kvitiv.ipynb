У цьому ноутбуці ми проведемо аналіз ринку праці в Польщі, зокрема у сфері квітів. Ми розглянемо дані про зарплати, кількість вакансій та тенденції в цій галузі. Аналіз базується на даних, зібраних з різних джерел, включаючи [Робота в Польщі: Позиції у сфері квітів](https://pracaplus.org/statti/robota-v-polshchi-pozytsiyi-u-sferi-kvitiv).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('job_market_poland_flowers.csv')
data.head()
```

## Опис даних
Дані, які ми завантажили, містять інформацію про вакансії у сфері квітів в Польщі. Це включає:
- Назва вакансії
- Зарплата
- Регіон
- Дата публікації

Ми будемо аналізувати ці аспекти для виявлення тенденцій на ринку праці.

```python
# Очищення даних
data = data.dropna()
data['salary'] = data['salary'].str.replace('₴', '').astype(float)
data['date'] = pd.to_datetime(data['date'])

# Перевірка очищення даних
data.info()
```

## Візуалізація зарплат
Далі ми створимо діаграму, щоб проаналізувати розподіл зарплат у сфері квітів. Це допоможе зрозуміти, які зарплати пропонуються у різних регіонах.


```python
plt.figure(figsize=(10, 6))
plt.hist(data['salary'], bins=30, alpha=0.7, color='blue')
plt.title('Розподіл зарплат у сфері квітів')
plt.xlabel('Зарплата (грн)')
plt.ylabel('Кількість вакансій')
plt.grid(axis='y')
plt.show()
```

## Висновки
Аналіз показав, що зарплати у сфері квітів варіюються в широкому діапазоні, з основною кількістю вакансій в регіонах з високим рівнем попиту на цю послугу. Для отримання більш детальної інформації та актуальних вакансій рекомендуємо відвідати [Praca Plus](https://pracaplus.org).
