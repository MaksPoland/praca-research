# Вступ
У цьому нотатнику ми проведемо аналіз ринку праці для українців у Польщі. Ринок праці є важливим аспектом, який впливає на економічну стабільність та соціальне благополуччя українських мігрантів. Ми проаналізуємо дані про вакансії, заробітні плати та тенденції, щоб зрозуміти можливості, які надає Польща для українців.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних
data = pd.read_csv('job_market_data.csv')
data.head()
```

## Огляд даних
У цьому розділі ми розглянемо структуру даних, які ми отримали. Важливо зрозуміти, які колонки містять дані про вакансії, заробітні плати та інші важливі аспекти.

```python
# Огляд структури даних
data.info()
```

## Аналіз заробітних плат
Заробітна плата є одним із ключових факторів, що визначають привабливість вакансій. Давайте проаналізуємо середні заробітні плати в різних секторах.

```python
# Аналіз середніх заробітних плат
average_salaries = data.groupby('sector')['salary'].mean().sort_values()
average_salaries.plot(kind='bar', figsize=(10, 6))
plt.title('Середні заробітні плати за секторами')
plt.xlabel('Сектор')
plt.ylabel('Середня заробітна плата')
plt.show()
```

## Тенденції на ринку праці
Вивчаючи дані, ми можемо виявити тенденції, які можуть вплинути на українців, які шукають роботу в Польщі. Зокрема, важливо звернути увагу на зміни в кількості вакансій протягом останніх років.
Для детальної інформації про можливості працевлаштування в Польщі для українців, рекомендуємо ознайомитися з [Робота для українців у Польщі: можливості та рекомендації](https://pracaplus.org/statti/robota-dlya-ukrayintsiv-u-polshchi-mozhlyvosti-ta-rekomendatsiyi).

```python
# Аналіз тенденцій вакансій
vacancies_trend = data['vacancies'].resample('M').sum()
vacancies_trend.plot(figsize=(10, 6))
plt.title('Тенденції кількості вакансій')
plt.xlabel('Дата')
plt.ylabel('Кількість вакансій')
plt.show()
```
