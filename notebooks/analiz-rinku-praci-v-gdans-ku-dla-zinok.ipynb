# Вступ
Цей ноутбук присвячений аналізу ринку праці в Гданську, зосередженому на можливостях для жінок. Ми розглянемо дані про вакансії, заробітні плати та тенденції на ринку праці, щоб визначити, які професії переважають і які можливості можуть бути доступні для жінок у цьому регіоні.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Завантаження даних
data = pd.read_csv('job_market_gdansk.csv')
data.head()
```

## Огляд даних
Ми завантажили дані про ринок праці в Гданську. У даних представлені різні категорії вакансій, розміри заробітних плат, а також інформація про те, скільки жінок працює в кожній категорії. Давайте розглянемо загальні характеристики даних.

```python
# Огляд статистики
data.describe()
```

## Вакансії та зарплати
На основі даних ми можемо проаналізувати, які вакансії є найбільш популярними серед жінок в Гданську. Також ми визначимо середні заробітні плати в різних галузях, щоб зрозуміти, де жінки можуть отримати найвищі доходи. Для детальнішого аналізу скористаємося графіками.
Джерело: [Робота в Гданську для жінок: можливості та рекомендації](https://pracaplus.org/statti/robota-v-hdansku-dlya-zhinok-mozhlyvosti-ta-rekomendatsiyi)

```python
# Групування даних за категоріями
category_salary = data.groupby('category')['salary'].mean() 
category_salary.plot(kind='bar', title='Середня зарплата за категоріями')
plt.ylabel('Зарплата (PLN)')
plt.show()
```

## Тенденції на ринку праці
Зараз ми розглянемо тенденції на ринку праці в Гданську. Слід проаналізувати, як змінилося попит на вакансії протягом часу, а також чи є певні галузі, які демонструють зростання.
Більше інформації можна знайти на основному сайті: [Praca Plus](https://pracaplus.org)

```python
# Аналіз тенденцій
trend_data = data.groupby('date')['vacancies'].sum() 
trend_data.plot(title='Тенденції вакансій у часі')
plt.ylabel('Кількість вакансій')
plt.show()
```

## Висновки
Аналіз ринку праці в Гданську показав, що жінки мають багато можливостей у різних галузях. Середні заробітні плати варіюються в залежності від вакансії, але є значні можливості для кар'єрного зростання. Сподіваємось, що цей аналіз стане корисним для жінок, які шукають роботу в Гданську.
