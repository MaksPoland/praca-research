Дослідження буде базуватись на відкритих даних про вакансії, зарплати та ринкові тенденції. Для більш детальної інформації, ви можете ознайомитись із статтею: [Робота в Польщі для жінок з дітьми: можливості та поради](https://pracaplus.org/statti/robota-v-polshchi-dlya-zhinok-z-ditmy-mozhlyvosti-ta-porady).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('job_market_poland.csv')
data.head()
```

## Огляд даних
Ми розглянемо дані про вакансії, включаючи назви посад, середні зарплати, а також регіони, в яких вони доступні. Це дозволить зрозуміти, які професії найпопулярніші серед жінок з дітьми.

```python
# Огляд статистики
data.describe()
```

## Аналіз зарплат
Зараз ми проаналізуємо середні зарплати для різних професій в Польщі. Це допоможе зрозуміти, які вакансії є найбільш вигідними.

```python
# Групування даних за професією та обчислення середньої зарплати
average_salaries = data.groupby('profession')['salary'].mean().sort_values(ascending=False)
average_salaries.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Середні зарплати за професіями')
plt.xlabel('Професії')
plt.ylabel('Середня зарплата')
plt.xticks(rotation=45)
plt.show()
```

## Висновки
З проведеного аналізу видно, що жінки з дітьми можуть знайти численні можливості для працевлаштування в Польщі. Однак важливо звертати увагу на середні зарплати та обирати професії, які найбільше підходять для їх потреб.
