# –í—Å—Ç—É–ø
–¶–µ–π –Ω–æ—É—Ç–±—É–∫ –ø—Ä–∏—Å–≤—è—á–µ–Ω–∏–π –∞–Ω–∞–ª—ñ–∑—É —Ä–∏–Ω–∫—É –ø—Ä–∞—Ü—ñ –≤ –ü–æ–ª—å—â—ñ –¥–ª—è —Å—ñ–º–µ–π–Ω–∏—Ö –ø–∞—Ä. –ú–∏ —Ä–æ–∑–≥–ª—è–Ω–µ–º–æ –¥–∞–Ω—ñ –ø—Ä–æ –∑–∞—Ä–ø–ª–∞—Ç–∏, –≤–∞–∫–∞–Ω—Å—ñ—ó —Ç–∞ —Ç–µ–Ω–¥–µ–Ω—Ü—ñ—ó, —è–∫—ñ –º–æ–∂—É—Ç—å –±—É—Ç–∏ –∫–æ—Ä–∏—Å–Ω–∏–º–∏ –¥–ª—è —É–∫—Ä–∞—ó–Ω—Å—å–∫–∏—Ö —Å—ñ–º–µ–π–Ω–∏—Ö –ø–∞—Ä, —è–∫—ñ –ø–ª–∞–Ω—É—é—Ç—å –ø—Ä–∞—Ü—é–≤–∞—Ç–∏ –≤ –ü–æ–ª—å—â—ñ.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –¥–∞–Ω–∏—Ö –ø—Ä–æ –≤–∞–∫–∞–Ω—Å—ñ—ó
data = pd.read_csv('job_market_poland.csv')
data.head()
```

## –û–≥–ª—è–¥ –¥–∞–Ω–∏—Ö
–£ —Ç–∞–±–ª–∏—Ü—ñ –≤–∏—â–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ñ –¥–∞–Ω—ñ –ø—Ä–æ –≤–∞–∫–∞–Ω—Å—ñ—ó, —è–∫—ñ –≤–∫–ª—é—á–∞—é—Ç—å —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—é –ø—Ä–æ –Ω–∞–∑–≤—É —Ä–æ–±–æ—Ç–∏, –∑–∞—Ä–ø–ª–∞—Ç—É, –º—ñ—Å—Ü–µ–∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è —Ç–∞ –¥–∞—Ç—É –ø—É–±–ª—ñ–∫–∞—Ü—ñ—ó. –¶—ñ –¥–∞–Ω—ñ –¥–æ–∑–≤–æ–ª—è—Ç—å –Ω–∞–º –∑—Ä–æ–∑—É–º—ñ—Ç–∏, —è–∫—ñ –≤–∞–∫–∞–Ω—Å—ñ—ó –Ω–∞–π–±—ñ–ª—å—à –ø–æ–ø—É–ª—è—Ä–Ω—ñ —Å–µ—Ä–µ–¥ —Å—ñ–º–µ–π–Ω–∏—Ö –ø–∞—Ä.

```python
# –§—ñ–ª—å—Ç—Ä–∞—Ü—ñ—è –¥–∞–Ω–∏—Ö –∑–∞ –∫–∞—Ç–µ–≥–æ—Ä—ñ—î—é –≤–∞–∫–∞–Ω—Å—ñ–π
filtered_data = data[data['category'] == '—Å—ñ–º–µ–π–Ω—ñ –ø–∞—Ä–∏']
filtered_data.describe()
```

## –ê–Ω–∞–ª—ñ–∑ –∑–∞—Ä–ø–ª–∞—Ç
–ó–∞—Ä–ø–ª–∞—Ç–∞ —î –æ–¥–Ω–∏–º –∑ –∫–ª—é—á–æ–≤–∏—Ö —Ñ–∞–∫—Ç–æ—Ä—ñ–≤, —è–∫—ñ –≤–ø–ª–∏–≤–∞—é—Ç—å –Ω–∞ –≤–∏–±—ñ—Ä —Ä–æ–±–æ—Ç–∏. –£ –Ω–∞—Å—Ç—É–ø–Ω–æ–º—É –∫–æ–¥—ñ –º–∏ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É—î–º–æ —Å–µ—Ä–µ–¥–Ω—é –∑–∞—Ä–ø–ª–∞—Ç—É –ø–æ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è–º –≤–∞–∫–∞–Ω—Å—ñ–π –¥–ª—è —Å—ñ–º–µ–π–Ω–∏—Ö –ø–∞—Ä.

```python
# –ê–Ω–∞–ª—ñ–∑ —Å–µ—Ä–µ–¥–Ω—å–æ—ó –∑–∞—Ä–ø–ª–∞—Ç–∏
average_salary = filtered_data.groupby('position')['salary'].mean().sort_values(ascending=False)
average_salary.plot(kind='bar', figsize=(12, 6), title='–°–µ—Ä–µ–¥–Ω—è –∑–∞—Ä–ø–ª–∞—Ç–∞ –ø–æ –≤–∞–∫–∞–Ω—Å—ñ—è—Ö –¥–ª—è —Å—ñ–º–µ–π–Ω–∏—Ö –ø–∞—Ä')
plt.ylabel('–ó–∞—Ä–ø–ª–∞—Ç–∞ (PLN)')
plt.xticks(rotation=45)
plt.show()
```

## –¢–µ–Ω–¥–µ–Ω—Ü—ñ—ó –Ω–∞ —Ä–∏–Ω–∫—É –ø—Ä–∞—Ü—ñ
–ó–≥—ñ–¥–Ω–æ –∑ –¥–æ—Å–ª—ñ–¥–∂–µ–Ω–Ω—è–º, –≤–∞–∫–∞–Ω—Å—ñ—ó –¥–ª—è —Å—ñ–º–µ–π–Ω–∏—Ö –ø–∞—Ä –≤ –ü–æ–ª—å—â—ñ —Å—Ç–∞—é—Ç—å –≤—Å–µ –ø–æ–ø—É–ª—è—Ä–Ω—ñ—à–∏–º–∏. –ë–∞–≥–∞—Ç–æ —Ä–æ–±–æ—Ç–æ–¥–∞–≤—Ü—ñ–≤ –ø—Ä–æ–ø–æ–Ω—É—é—Ç—å –≥–Ω—É—á–∫—ñ —É–º–æ–≤–∏ –ø—Ä–∞—Ü—ñ, —â–æ —Ä–æ–±–∏—Ç—å —ó—Ö –ø—Ä–∏–≤–∞–±–ª–∏–≤–∏–º–∏ –¥–ª—è —É–∫—Ä–∞—ó–Ω—Å—å–∫–∏—Ö —Å—ñ–º–µ–π. –î–ª—è –±—ñ–ª—å—à –¥–µ—Ç–∞–ª—å–Ω–æ–≥–æ –∞–Ω–∞–ª—ñ–∑—É, –¥–∏–≤—ñ—Ç—å—Å—è —Å—Ç–∞—Ç—Ç—é: [–†–æ–±–æ—Ç–∞ –≤ –ü–æ–ª—å—â—ñ –¥–ª—è —Å—ñ–º–µ–π–Ω–∏—Ö –ø–∞—Ä: –º–æ–∂–ª–∏–≤–æ—Å—Ç—ñ —Ç–∞ –ø–æ—Ä–∞–¥–∏](https://pracaplus.org/article.php?id=2a57cb53-e4d5-4743-8285-2cd0269fb04a).

```python
# –í—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è —Ç–µ–Ω–¥–µ–Ω—Ü—ñ–π
vacancy_trends = data['date_posted'].value_counts().sort_index()
vacancy_trends.plot(kind='line', figsize=(12, 6), title='–¢–µ–Ω–¥–µ–Ω—Ü—ñ—ó –≤–∞–∫–∞–Ω—Å—ñ–π —É –ü–æ–ª—å—â—ñ')
plt.ylabel('–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∞–∫–∞–Ω—Å—ñ–π')
plt.xlabel('–î–∞—Ç–∞')
plt.show()
```

–ê–Ω–∞–ª—ñ–∑ —Ä–∏–Ω–∫—É –ø—Ä–∞—Ü—ñ –≤ –ü–æ–ª—å—â—ñ –ø–æ–∫–∞–∑—É—î, —â–æ —ñ—Å–Ω—É—î –±–µ–∑–ª—ñ—á –º–æ–∂–ª–∏–≤–æ—Å—Ç–µ–π –¥–ª—è —É–∫—Ä–∞—ó–Ω—Å—å–∫–∏—Ö —Å—ñ–º–µ–π–Ω–∏—Ö –ø–∞—Ä. –†–µ–∫–æ–º–µ–Ω–¥—É—î–º–æ –∑–≤–µ—Ä—Ç–∞—Ç–∏—Å—è –¥–æ –≤–µ–±-—Å–∞–π—Ç—ñ–≤, —è–∫—ñ —Å–ø–µ—Ü—ñ–∞–ª—ñ–∑—É—é—Ç—å—Å—è –Ω–∞ –ø—Ä–∞—Ü–µ–≤–ª–∞—à—Ç—É–≤–∞–Ω–Ω—ñ, —Ç–∞–∫–∏—Ö —è–∫ [Praca Plus](https://pracaplus.org), –¥–ª—è –æ—Ç—Ä–∏–º–∞–Ω–Ω—è –∞–∫—Ç—É–∞–ª—å–Ω–æ—ó —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó.
