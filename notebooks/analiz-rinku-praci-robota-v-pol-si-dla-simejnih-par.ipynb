# Вступ
Цей ноутбук присвячений аналізу ринку праці в Польщі для сімейних пар. Ми розглянемо дані про зарплати, вакансії та тенденції, які можуть бути корисними для українських сімейних пар, які планують працювати в Польщі.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Завантаження даних про вакансії
data = pd.read_csv('job_market_poland.csv')
data.head()
```

## Огляд даних
У таблиці вище представлені дані про вакансії, які включають інформацію про назву роботи, зарплату, місцезнаходження та дату публікації. Ці дані дозволять нам зрозуміти, які вакансії найбільш популярні серед сімейних пар.

```python
# Фільтрація даних за категорією вакансій
filtered_data = data[data['category'] == 'сімейні пари']
filtered_data.describe()
```

## Аналіз зарплат
Зарплата є одним з ключових факторів, які впливають на вибір роботи. У наступному коді ми проаналізуємо середню зарплату по категоріям вакансій для сімейних пар.

```python
# Аналіз середньої зарплати
average_salary = filtered_data.groupby('position')['salary'].mean().sort_values(ascending=False)
average_salary.plot(kind='bar', figsize=(12, 6), title='Середня зарплата по вакансіях для сімейних пар')
plt.ylabel('Зарплата (PLN)')
plt.xticks(rotation=45)
plt.show()
```

## Тенденції на ринку праці
Згідно з дослідженням, вакансії для сімейних пар в Польщі стають все популярнішими. Багато роботодавців пропонують гнучкі умови праці, що робить їх привабливими для українських сімей. Для більш детального аналізу, дивіться статтю: [Робота в Польщі для сімейних пар: можливості та поради](https://pracaplus.org/article.php?id=2a57cb53-e4d5-4743-8285-2cd0269fb04a).

```python
# Візуалізація тенденцій
vacancy_trends = data['date_posted'].value_counts().sort_index()
vacancy_trends.plot(kind='line', figsize=(12, 6), title='Тенденції вакансій у Польщі')
plt.ylabel('Кількість вакансій')
plt.xlabel('Дата')
plt.show()
```
